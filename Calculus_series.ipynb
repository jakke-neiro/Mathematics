\section{Seriens konvergens}
\subsection{Definitioner}
$x_n$ talföljd, $s_n$ n:te delsumman, $s_1 = x_1$, $s_2 = x_1 + x_2$, $s_n = \sum_{k=1}^n x_k = x_1 + x_2 + x_3 + \dots + x_2, n \in \N_1$

Serien: $\sum_{k=1}^{\infty}x_k = x_1 + x_2 + x_3 + \dots$

Serien konvergerar: $\exists s \in \R, \lim_{n \to \infty}s_n = s$, i annat fall divergerar serien

$r_n= s - s_n$ n:te resttermen

Aritmetisk serie: $\sum x_k$ är aritmetisk, om $\exists d \in \R$: $x_{k+1} - x_k = d \forall k \in \N_1$
\[
n\frac{a_1 + a_n}{2}
\]

Geometrisk serie: $\sum x_k$ är geometrisk, om $\exists q \in \R:$
$x_{k+1} = qx_k \forall k \in \n_1$
\[
a_1 \frac{1-q^n}{1-q}
\]

Geometrisk serie: $\sum_{k=0}^{\infty}aq^k$, $a \neq 0$, 1. konvergerar $|q| < 1 \to \frac{a}{1-q}$, 2.divergerar $|q| \geq 1$

\subsection{Grundsatser}
Om serien $\sum x_k$ konvergerar, så $\lim_{k \to \infty}x_k = 0$ och $\lim_{k \to \infty}r_k = 0$

Låt $p \in \N_0$. Serien $\sum_{k=p+1}^{\infty}x_k$ konvergerar, omm $\sum_{k=1}^{\infty}x_k$ konvergerar, alltså $\sum_{k=1}^{\infty}x_k = \sum_{k=1}x_k^p + \sum_{k=p+1}^{\infty}x_k$

$\sum x_k + \sum y_k = \sum_{k=1}^{\infty}(x_k + y_k)$, $c\sum x_k = \sum c x_k$, $\sum_{k=1}^{\infty}(x_k + y_k) = s + t$, $\sum_{k=1}^{\infty}c x_k = cs$

Cauchys allmänna konvergenskriterium:

Serien $\sum x_k$ konvergerar, omm $\forall \epsilon > 0$ motsvarar $n_{\epsilon} \in \N_1$, så att 
\[
\bigg| \sum_{k=n+1}^{n+p}x_k \bigg| < \epsilon \forall p \in \N_1, n > n_{\epsilon}
\]

\subsection{Majorant- och minorantprincipen, p-serier}
Icke-negativ $x_k \geq 0$, positiv $x_k > 0$, icke-positiv $x_k \leq 0$, negativ $n_k < 0$.

Delsumman: talföljden av delsummor $s_n$ är begränsad från ovan

Integraltestet: $\sum_{k=n}^{\infty}x_k$ icke-negativ, $n \in \N_1$. Anta att $f:[n, \infty) \to \R$ är avtagande, $x_k = f(k) \forall k \geq n, k \in \N$, och $f$ är integrerbar p alla intervall $[n, b]$ var $b\in \R$ och $b > n$. Då konvergerar serien endast om $\int_n^{\infty} f(x) dx$ konvergerar

$p\in \R$. Serien $\sum_{k=1}^{\infty}\frac{1}{k^p}$ är en p-serie 1. $p < 1$ subharmonisk, 2. $p=1$ harmonisk, 3. $p > 1$ superharmonisk. p-serien $\sum_{k=1}^{\infty}\frac{1}{k^p}$ konvergerar, då $p > 1$

Majorantprincipen: $\sum y_k$ konvergerar, så $\sum x_k$ konvergerar och $\sum x_k \leq \sum y_k$

Minorantprincipen: $\sum x_k$ divergerar, så $\sum y_k$ divergerar

\subsection{Konvergenstest}
Jämförelsetestet: $\sum x_k$ icke-negativ och $\sum y_k$ positiv
\begin{enumerate}
    \item $\lim_{k \to \infty}\frac{x_k}{y_k} = c > 0$ $x$ konvergerar omm $y$ konvergerar
    \item $\lim_{k \to \infty}\frac{x_k}{y_k}$
    \item $\lim_{k \to \infty}\frac{x_k}{y_k} = \infty$ och $y_k$ divergar $\to$ $x_k$ divergerar
\end{enumerate}

Kvottestet: $\sum x_k$ positiv
\begin{enumerate}
    \item $\lim_{k \to \infty}\frac{x_{k+1}}{x_k} = c < 1$
    $\to$ $x$ konvergerar
    \item $\lim_{k \to \infty}\frac{x_{k+1}}{x_k} = c > 1$ eller $\lim_{k \to \infty}\frac{x_{k+1}}{x_k} = \infty$ $\to$ divergerar
\end{enumerate}

Rottestet: $\sum x_k$ icke-negativ
\begin{enumerate}
    \item $\lim_{k \to \infty}\sqrt[k]{x_k} = c < 1$ $\to$ konvergerar
    \item $\lim_{k \to \infty}\sqrt[k]{x_k} = c > 1$ eller $\lim_{k \to \infty}\sqrt[k]{x_k} = \infty$ $\to$ divergerar
\end{enumerate}

\section{Alternerande och manipulerade serier}
\subsection{Alternerande serier och absolut konvergens}
Serien $\sum x_k$ är alternerande, om dess termer är turvis icke-negativa och icke-positiva
\[
\sum_{k=1}^{\infty}(-1)^{k-1}x_k, x_k \geq 0 \forall k \in \N_1
\]
eller
\[
\sum_{k=1}^{\infty}(-1)^k x_k, x_k \geq 0 \forall k \in \N_1
\]

Leibniz test: Anta att $x_k$ är icke-negativa $\forall k \in \N_1$, $x_k \geq x_{k+1}$, $\lim_{k \to \infty} x_k = 0$. Då konvergerar de alterande serierna $\sum_{k=1}^{\infty}(-1)^k x_k$. Om antagandena för Leibniz test gäller, då gäller att 
\[
|s - s_n| = |r_n| \leq x_{n+1} \forall n \in \N_1
\]

Serien $\sum_k$ konvergerar absolut, om serien $\sum |x_k|$ konvergerar.

Om serien $\sum x_k$ konvergerar abolut, så konvergerar den. Dessutom gäller att 
\[
\bigg| \sum_{k=1}^{\infty}x_k \bigg| \leq \sum_{k=1}^{\infty}|x_k|
\]

Om serien $\sum x_k$ divergerar, så divergerar den absolut

\subsection{Organisering av termer och villkorlig konvergens}
Låt $n_1 < n_2 < n_3 < \dots < n_k \in \N_1 \forall k \in \N_1$. Då är 
\[
\sum_{k=1}^{\infty}\tilde{x}_k = (x_1 + \dots + x_{n_1}) + (x_{n_1 + 1} + \dots + x_{n_2}) + (x_{n_2 + 1} + \dots + x_{n_3}) + \dots
\]
Om serien $\sum x_k$ konvergerar, så konvergerar $\sum \tilde{x}_k$ och $s_1 = s_2$.

Om serien $\sum x_k$ är icke-negativ och $\sum_{k=1}^{\infty}\tilde{x}_k = (x_1 + x_2) + (x_3 + x_4) + \dots $ konvergerar $\to$ $\sum x_k$ konvergerar, om $\lim_{k \to \infty}x_k = 0$

Serien $\sum x_k$ konvergerar villkorligt, om den konvergerar men divergerar absolut.

Serien $\sum x_k$ konvergerar absolut, omm $\sum x_k ^+$ och $\sum x_k ^-$ konvergerar.

Om serien $\sum x_k$ konvergerar villkorligt, så divergerar $\sum x_k^+$ och $\sum y_k^-$

\subsection{Omorganisering av termer}
Låt $\phi: \N_1 \to \N_1$ vara en bijektion. Då är 
\[
\sum_{k=1}^{\infty}x_{\phi(k)} = x_{\phi(1)} + x_{\phi(2)} + x_{\phi(3)} + \infty
\]

Anta att den icke-negativa serien $\sum x_k$ konvergerar och dess summa är $s$. Då konvergerar $\sum x_{\phi(k)}$ och summan är $s$.

Riemanns omorganiserignstest: Anta att serien $\sum x_k$ konvergerar villkorligt och $s \in \R$. Då $\exists$ bijektionerna $\phi, \psi: \N_1 \to \N_1$, så att $\sum x_{\phi}(k)$ konvergerar och dess summa är $s$. $\sum x_{\psi}(k)$ divergerar. 

\section{Potensserier}
\subsectjon{Definjtion och Abels sats}
Potensserier är serierns $\sum_{k=0}^{\infty}a_k(x-x_0)^k$, $x \in \R$, där $a_0, a_1, a_2, \dots, \in \R$ är seriens koefficienter och $x_0 \in R$ är seriens centrum.

Abels sats: Om potensserien $\sum a_k (x-x_0)^k$ konvergerar på värder $x=y$, $y \neq x_0$, så konvergerar serien $\forall x \in \R: |x - x_0| < |y - x_0|$

Om potensserien $\sum a_k(x - x_0)^k$ divergerar absolut på värder $x = z$, så divergerar serien $\forall x \in \R: |x-x_0| > |z - x_0|$

Potensseriens $\sum a_k(x-x_0)^k$ konvergensintervall är $\{x \in \R$: serien konvergerar $\}$ och konvergensradien är $\sup \{|x-x_0|:$ serien konvergerar $\}$. 

Låt $\sum a_k (x - x_0)^k$ vara en potensserie och $R$ dess konvergensradie
\begin{enumerate}
    \item Serien konvergerar då $x = x_0$, omm $R = 0$
    \item Serien konvergerar absolut $\forall x$, omm $R = \infty$
    \item Seriens $m$ är ett positivt $\R$, så konvergerar serien absolut $\forall x \in (x_0 - m, x_0 + m)$ och divergerar $\forall x \notin [x_0 - m, x_0 + m]$, omm $R = m$
\end{enumerate}

\subsection{Konvergensradien}
Låt $\sum a_k(x-x_0)^k$ vara en potensserie, för koefficienterna $n \in \N_0$ $a_k \neq 0 \forall k \geq n, k \in \N_0$. Om gränsvärdet $\lim_{k \to \infty}\bigg|\frac{a_k}{a_{k+1}}\bigg|$ eller $\lim_{k \to \infty}\frac{1}{\sqrt[k]{|a_k|}} \exists$, så är potensseriens $\sum a_k (x - x_0)^k$ konvergensradie $\R$.

Potensserien $\sum_{k=1}^{\infty}ka_k(x-x_0)^{k-1}$ och $\sum_{k=0}^{\infty}\frac{a_k}{k+1})(x-x_0)^{k+1}$ har samma konvergensradie som potensserien $\sum_{k=0}^{\infty}a_k (x - x_0)^k$.

För potensseriens $\sum a_k (x - x_0)^k$ konvergensrqadie $R$ gäller:
\begin{enumerate}
    \item Om det $\exists m > 0$ att $|a_k| \geq m \forall k \in \N_0 \to \R \leq 1$
    \item Om det $\exists M < \infty$ att $|a_k| \leq M \forall k \in \N_0 \to R \geq 1$
    \item Om båda gäller $\to R = 1$
\end{enumerate}

\subsection{Seriens likformiga konvergens}
Låt $A \subset \R$ och $f_k: A \to \R, k \in \N_1$. Då kan delsumsfunktionen $s_n: A \to \R$ $s_n(x) = \sum_{k=1}^n f_k(x), x \in A$. Serien $\sum f_k$ konvergerar punktvis i mängden $A \subset \R$ mot gränsfunktionen $s: A \to \R$, om serien $\sum_{k=1}^{\infty}f_k(x)$ konvergerar mot $s(x) \forall x \in A$.

Serien $\sum f_k$ konvergerar likformigt i mängden $A \subset \R$ mot gränsfunktionen $S: A \to \R$, om $\forall \epsilon > 0$ $\exists n_0 \in \N_1$, $\sup_{x \in A}|\sum_{k=1}^n f_k(x) - s(x)| \leq \epsilon, n > n_0$

Weierstrass M-test: $A \subset \R$ och $f_k: A \to \R$, $k \in \N_1$. Om $M_k \in \R$, så att $|f_k(x)|\leq M_k \forall x \in A$ och $k \in \n_1$ och serien $\sum M_k$ konvergerar, så konvergerar $\sum f_k$ likformigt i mängden $A$.

Om potensseriens $\sum a_k(x- x_0)^k$ konvergensradie är $R > 0$, så konvergerar den likformigt på varje slutet intervall $[x_0 - 0, x_0 + p]$, $0 < p < R$.  $A \subset R$, om funktionerna $f_k: A \to \R$, $k \in \N_1$ är kontinuerliga i $A$ och $\sum f_k$ konvergerar absolut i $A$, $\to \sum f_k: A \to \R$ är kontinuerlig i $A$.

Korollarium: Om potensseriens $\sum a_x_0)^k$ konvergensradie $R > 0$, så är dess summafunktion $s(x) = \sum_{k=0}^{+infty}a_k(x-x_0)^k$ kontinuerlig i $(x_0-R, x_0 + R)$

Anta att funktionerna $f_k:[a, b] \to \R, k \in \N_1$ är integrerbara och att serien $\sum f_k$ konvergerar likformigt i intervallet $[a, b]$. Då är
\[
\int_a^b \bigg(\sum_{k=1}^{\infty}f_k(x) \bigg)dx = \sum_{k=1}^{\infty} \int_a^b f_k(x) dx
\]

Korollarium: Potensseriens $\sum_{k=0}^{\infty}a_k(x-x_0)^k$ summafunktion
\[
s(x) = \sum_{k=0}^{\infty}a_k(x-x_0)^k, x \in (x_0 - R, x_0 + R)
\]
$R$ konvergensradie, kan inegreras termvis $\int_{x_0}^x s(t) dt = \sum_{k=0}^{\infty}\frac{a_k}{k+1}(x-x_0)^{k+1}$

Anta att $\Delta \subset \R$ är ett slutet intervall, funktionerna $f_k: \Delta \to \R$, $k \in \N$, är kontinuerligt deriverbara i intervallet $\Delta$ och serierna $\sum f_k$ och $\sum f_k'$ konvergerar likformigt i intervallet $\Delta$. Då är summafunktionen $s = \sum f_k: \Delta \to \R$ kontinuerligt deriverbar på intervaller $\Delta$ och $s'(x) = \sum f_k'(x) \forall x \in \Delta$.

Korollarium: Potensseriens $\sum_{k=0}^{\infty}a_k(x-x_0)^k$ summafunktion
\[
s(x) = \sum_{k=0}^{\infty}a_k(x - x_0)^k, x \in (x_0 - R, x_0 + R)
\]
kan deriveras termvis $s'(x) = \sum_{k=1}^{\infty}ka_1(x -x_0)^{k-1}$ $\forall x \in (x_0  R, x_0 + R)$-

$S(x_0) = a_0$ och $s^{(n)}(x_0) = n!a_n$

Potensseriens entydighet: Om potensserierna $\sum a_k (x - x_0)^k$ och $\sum b_k (x - x_0)^k$ konvergerar och uppvisar samma funktion någonstans i ett öppet intervall innehållande $x_0$, så $a_k = b_k \forall k \in \N_0$.

\section{Taylors och Maclaurins polynom}
\subsection{Taylors formel}
Låt $x_0 \in \R$ och $h > 0$. Anta att $\exists f: (x_0-h, x_0 + h) \to \R$, $\exists f'(x), f''(x), \dots, f^{(n-1)}(x) \forall x \in (x_0 - h, x_0 + h)$ och $\exists f^{(n)}(x_0)$. Då är polynomfunktionen $T_n(x; x_0): \R \to \R$
\[
T_n(x; x_0) = f(x_0) + \sum_{k=1}^n \frac{f^{(k)}(x_0)}{k!}(x - x_0)^k, n \in \N_0
\]
$f \in C^{\infty}(\Delta)$: funktionen $f$ kan deriveras oändligt många gånger och alla derivator är kontinuerliga i det öppna intervallet $\Delta$.

Taylors formel: Låt $x_0 \in \R$ och $n \in \N_0$. Om $f \in C^{n+1}(x_0-h, x_0+h)$ för något $h > 0$, så $\forall x \in (x_0 - h, x_0 + h)$ är $f(x) = T_n(x; x_0) + R_{n+1}(x; x_0)$, där restermen $R_{n+1}(x; x_0) = \int_{x_0}^x \frac{f^{(n+1)}(t)}{n!}(x-t)^ndt$

\begin{enumerate}
    \item Resttermsformeln: $R_{n+1}(x; x_0)$ kan $\forall x \in (x_0-h, x_0+h)$ skrivas $(x-x_0)^n\epsilon(x)$, var $\epsilon: (x_0-h, x_0 + h) \to \R$ är en funktion, för vilken $\lim_{x \to x_0}\epsilon(x) = \epsilon(x_0) = 0$
    \item Cauchys restterm: $\forall x \in (x_0 -h, x_0 + h)$ är 
    \[
    R_{n+1}(x; x_0) = \frac{f^{(n+1)}(\epsilon_x)}{n!}(x-\epsilon_x)^n(x-x_0)
    \]
    \item Lagranges restterm: $\forall x \in (x_0 - h, x_0 + h)$ är 
    \[
    R_{n+1}(x; x_0) = \frac{f^{(n+1)}(\epsilon_x)}{(n+1)!}(x - x_0)^{n+1}
    \]
\end{enumerate}

\subsection{Entydighet}
Entydighetem för Taylors utveckling: Låt $x_0 \in \R$ och $f \in C^{n+1}(x_0-h, x_0+h)$ för någpt $h>0$. 
\[
f(x) = a_0 + a_1(x -x_0) + \dots a_n(x - x_0)^n + (x - x_0)^n\epsilon_1(x)
\]
\[
f(x) = b_0 + b_1(x - x_0) + \dots + b_n(x-x_0)^n + (x-x_0)^n \epsilon_2(x)
\]
$\forall x \in (x_0 -h, x_0+h)$ där $\lim_{x \to x_0}\epsilon_j(x) = 0$, $j = 1, 2,$ så är $a_k = b_k \forall x \in (x_0 -h, x_0+h)$ och således $\epsilon_1(x) = \epsilon_2(x) \forall x \in (x_0 - h, x_0 + h)$.

Korollarium: Låt $x_0 \in \R$ och $f \in C^{n+1}(x_0 - h, x_0 + h)$ för något $h>0$. Om funktionen $f$ har utvecklingen 
\[
f(x) = a_0 + a_1(x - x_0) + \dots + a_n(x_x_0)^n + (x-x_0)^n\epsilon(x)
\]
$\forall x \in (x_0-h, x_0 +h)$, var $\lim_{x \to x_0}\epsilon(x) = 0$, så är det en Taylor-utveckling.

Låt $f, g \in C^{n+1}(x_0-h, x_0 + h)$ för något $h>0$. Om $c \in \R$ och Taylorutvecklingarna för $f$ och $g$ är $f(x) = T_n^g(x; x_0) + R_{n+1}^f(x; x_0)$ och $g(x) = T_n^g(x; x_0) + R_{n+1}^g(x; x_0)$ så 
\[
f(x) + g(x) = (T_n^f(x; x_0) + T_n^g(x)(x;x_0)) + (R_{n+1}^f(x; x_0) + R_{n+1}^g(x; x_0))
\]
\[
(cf)(c) = cT_n^f(x; x_0) + cR_{n+1}^f(x, x_0)
\]

\subsection{Tillämpningar}
Gränsvärden och integraler

Låt $f \in C^n(x_0 - h, x_0 +h)$ och anta att $f'(x_0) = f''(x_0) = \dots = f^{(n-1)}(x_0) = 0$ och $f^{(n)}(x_0) \neq 0$. Då har funktionen $f$ i punkten $x_0$
\begin{enumerate}
    \item Ett lokalt maximum, om $n$ är jämnt och $f^{(n)}(x_0) < 0$
    \item Ett lokalt minimu, om $n$ är jämnt och $f^{(n)}(x_0) > 0$
    \item Inget extremvärde, om $n$ är udda
\end{enumerate}

\section{Taylors serie}
Låt $x_0 \in \R$, $h > 0$ och $f \in C^{\infty}(x_0 - h, x_0+h)$. Då är potensserien 
\[
\sum_{k=0}^{\infty}\frac{f^{(k)}(x_0)}{k!}(x - x_0)^k = f(x_0) + \sum_{k=1}^{\infty}\frac{f^{(k)}(x_0)}{k!}(x-x_0)^k
\]
Funktionens $f$ Taylorserie i $x_0$

Låt $f \in C^{\infty}(x_0 - h, x_0 + h), h>0$. Om $y \in (x_0-h, x_0 + h)$ och $\lim_{n \to \infty} R_n(y; x_0)$, så
\[
f(y) = \sum_{k=0}^{\infty}\frac{f^{(k)}(x_0)}{k!}(y - x_0)^k
\]
Funktionen $f \in C^{\infty}(x_0 - h, x_0 + h)$ är på intervallet $(x_0-r, x_0+r)$, $0 > r \leq h$, en summa tille en konvergerande potensserie $\sum_{k=0}^{\infty} a_k(x - x_0)^k$, omm
\[
\lim_{n \to \infty}\frac{f^{(n)}(\epsilon_{x, n}}{n!}(x - x_0)^n = 0 \forall x \in (x_0 - r, x_0 +r)
\]
Då är $\sum_{k=0}^{\infty}a_k(x_x_0)^k$ en Taylorserie till $f$ alltså $a_k = \frac{f^{(k)}(x_0)}{k!} \forall k$. Entydigheten för Taylors serie: Låt $x_0 \in \R$ och $f \in C^{\infty}(x_0-h, x_0 +h), h>0$. Om 
\[
f(x) = \sum_{k=0}^{\infty}a_k(x - x_0)^k, x \in (x_0-h, x_0 + h)
\]
så är denna utveckling funktionens Taylorserie.
