\section{Riemanns integral}
\subsection{Riemanns integral}
Delningen $J$ för intervallet $[a, b]$ är den ändliga följd $x_0, x_1, \dots, x_n$ för vilken $a = x_0 < x_1 < \dots < x_n-1 < x_n = b$

Delningens $J$ norm $|J = \textrm{max}\{x_k - x_{k-1}: k = 1, \dots, n \}$

$\Delta_k = [x_{k-1}, x_k]$, $p_k = \inf_{x \in \Delta_k} f(x)$ och $s_k = \sup_{x \in \Delta_k}f(x)$

Undersumman: $\underline{S}_J(f) = \sum_{k=1}^n p_k (x_k - x_{k-1}) = p_1(x_1 - x_0) + \dots + p_n(x_n - x_{n-1})$

Översumman: $\overline{S}_J (f) = \sum_{k=1}^n s_k(x_k - x_{k-1}) = s_1(x_1 - x_0) + \dots + s_n(x_n - x_{n-1})$
\[
\underline{S}_J(f) \leq \overline{S}_J(f)
\]
Delningen $A$ är en förlängning av $J$: $\underline{S}_J (f) \leq \underline{S}_A(f)$ och $\overline{S}_A (f) \leq \overline{S}_J(f)$

$A(f) = \sup_J \underline{S}_J (f) = \inf_J \overline{S}_J (f) = Y(f)$, alla delningar i intervallet $[a, b]$

\subsection{Riemanns integrerbarhetsvillkor}
För varje $\epsilon > 0$ finns en sådan delning $J$, 
\[
\overline{S}_J(f) - \underlin{S}_J (f) < \epsilon
\]
Monotona funktioner är integrerbara

Kontinuerliga funktioner i intervallet $[a, b]$ är integrerbara i $[a, b]$

Styckevist kontinuerliga funktioner är integrerbara

\subsection{Riemanns summor}
$f: [a, b] \to \R$ för alla $\epsilon > 0$ finns ett sådant $\elta > 0$, att $Y(f) \leq \overline{S}_J(f) < Y(f) + \epsilon$ och $A(f) - \epsilon < \underline{S}_J(f) \leq A(s)$ för alla delningar $J$, för vilka $|J| < \delta$.

$\lim_{n \to \infty}\overline{S}_{Jn}(f) = Y(f)$ och $\lim_{n \to \infty}\underline{S}_{Jn}(f) = A(f)$

$f:[a, b] \to \R$, delningen $J: [a, b]$ följden $z_1, z_2, \dots, z_n$ och $z_k \in [x_{k-1}, x_k]$ för alla tal $k=1, \dots, n$

$S_J(f, z_1, \dots, z_n) = \sum_{k=1}^n f(z_k)(x_k - x_{k-1}) = f(z_1)(x_1 - x_0) + \dots + f(z_n)(x_n - x_{n-1})$

$S_J(f, z_1, \dots, z_n) \to I \in \R$, om för vajre tal $\epsilon > 0$ finns ett sådant $\delta > 0$, att
\[
|S_J(f, z-1, \dots, z_n) _ I| < \epsilon
\]

\section{Integralfunktion}
\subsection{Inledning}
Låt $\Delta \subset \R$ vara ett intervall $f, F:\Delta \to \R$. Om $F$ är deriverbar på intervallet $\Delta$ och $F'(x) = f(x)$ på alla $x \in \Delta$, funktionen $F$ är funktionen $f$:s integralfunktion $F+c, c\in \R$.

Låt $c \in \R$. Funktionerna $f$ och $g$ har integralfunktionerna $F$ och $G$. Då har funktionerna $f+g$ och $cf$ de motsvarande integralfunktionerna $F+G$ och $cF$

\subsection{Riemanns integral och integralfunktionen}
$F:[a, b] \to \R$, $F(x) = \int_a^x f(t)dt$ kontinuerlig

$f$ är kontinuerlig i $x_0 \in (a, b)$, då $F:[a, b] \to \R$

$F(x) = \int_a^x f(t)dt$ är deriverbar i punkten $x_0$, och $F'(x_0) = f(x_0)$. Om $f$ är kontinuerlig i punkten $a$ eller $b$, så är $F$ deriverbar från höger och vänster och $F'(a) = f(a)$ och $F'(b) = f(b)$.

Analysens grundsats I: Funktionen $f$ är kontinuerlig på intervallet $[a, b]$. Då har den på intervallet $[a, b]$ en integralfunktion $F:[a, b] \to \R$, $F(x) = \int_a^x f(t)dt$, och alla integralfunktioner på intervallet $[a, b]$ är $F(x) + C, C \in \R$.

Analysens grundsats II: $f$ integrerbar $[a, b]$, $F$ är en integralfunktion, då är 
\[
\int_a^b f(x) dx = \bigg|_a^b F(x) = F(b) - F(a)
\]
\[
F(x) = \int_a^x f(t)dt + F(a), x \to \int_a^x f(t) dt
\]

\subsection{Integralkalkylens medelvärdessats}
Funktionen $f$ är kontinuerlig på intervallet $[a, b]$, så finns det en punkt $z \in (a, b)$, att
\[
\frac{1}{b-a}\int_a^b f(x) dx = f(z)

\section{Oegentlig integral}
\subsection{Definition}
Låt $f:[a, \infty) \to \R$ integrerbar på intervallet $[a, b] \subset [a, \infty)$. Låt $E: [a, \infty) \to \R$.

$E(x) = \int_a^x f(y) dy$. Funktionen $E$ är oegentlig interal $[a, \infty)$ och vi märker att $\int_a^{\infty}f(x)dx$

Om $\lim_{x \to \infty}E(x) = r \in \R$, så konvergerar den oegentliga integralen. Annars divergerar den oegentliga integralen.

Låt $f: [a, b] \to \R$ integrerbar på intervallet $[a, c] \subset [a, b)$. Låt $E: [a, b) \to \R$, $E(x) = \int_a^x f(y)dy$, funktionen $E$ är oegentlig integral $[a, b]$ och vi märker att $\int_a^b f(x) dx$.

Om $\lim_{x \to b^-}E(x) = r$, så konvergerar den oegentliga integralen. Annars divergerar den oegentliga integralen.

\subsection{Den oegentliga integralens egenskaper}
$f:[a, b] \to \R$ är integrerbar, gränsvärden
\[
\lim_{c \to b^-}\int_a^c f(x) dx = \lim_{c \to a^+}\int_a^b f(x) dx = \int_a^b f(x) dx
\]

\section{Substitution method}

\[
\int_{x_1}^{x_2}f(g(x))dx = \int_{u_1}^{u_2}f(u)\frac{dx}{du}du
\]

\begin{example}
\[
\int_a^b (2x + 3)^4 dx
\]
\[
u = 2x + 3 \iff x = \frac{1}{2}(u - 3) = \frac{u}{2} - \frac{3}{2} \iff \frac{dx}{du} = \frac{1}{2}
\]
Substitution
\[
\int_{u(a)}^{u(b)}\frac{1}{2}u^4du = \frac{1}{2}\int_{u(a)}^{u(b)}u^4 du = \frac{1}{2} \bigg[ \frac{1}{5}u^5 \bigg]_{u(a)}^{u(b)} = \frac{1}{2} \bigg[ \frac{1}{5}(2x + 3)^5 \bigg]_{a}^{b}
\]
\end{example}
