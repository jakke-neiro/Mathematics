\section{Reaaliarvoiset vektorifunktiot}
\subsection{Raja-arvo}
$A \subset \R^n$, $a$ joukon $A$ kasautumispiste. Kuvauksella $f:A \to \R$ on pisteessä $a$ raja-arvo $\b \in \R$ joukon $A$ suhteen, jos jokaiselle annetulle $\epsilon > 0$ on olemassa $\delta_{\alpha, \beta} = \delta > 0$ siten, että
\[
|f(y) - b| < \epsilon \textrm{ kun } y \in A \textrm{ ja } 0 < \| y - a \| < \delta 
\]
\[
\lim_{y \to a}f(y) = b
\]
\begin{example}
Onko raja-arvoa $\lim_{(x, y) \to (0, 0)}\frac{3x^2y}{x^2 + y^2}$?
\[
y = x^2, \frac{3x^2 y}{x^2 + y^2} = \frac{3x^4}{x^2 + x^4} = \frac{x^23x^2}{x^2(1 + x^2)} \to 0, x \to 0
\]
\[
x = y^2, \frac{3x^2y}{x^2 + y^2} = \frac{3y^5}{y^4 + y^2} = \frac{y^2 3 y^3}{y^2(y^2 + 1)} \to 0, y \tp 0
\]
Raja-arvo on ehkä olemassa!

$\epsilon > 0$, $0 < \|(x, y) - (0,0) \| < \delta$ ja $||(x, y) - (0,0)\| = \|(x, y)\| = \sqrt{x^2 + y^2}$

\[
\bigg|\frac{3x^2 y}{x^2 + y^2} - 0 \bigg| < \epsilon \iff \frac{3x^2|y|}{x^2 + y^2} < \epsilon 
\]
\[
\frac{3x^2|y|}{x^2 + y^2} \leq 3|y| = 3 \sqrt{y^2} \leq 3 \sqrt{x^2 + y^2} < 3 \delta \to \delta = \epsilon/3
\]
\end{example}

\subsection{Jatkuvuus}
Olkoon $A \subset \R^n$ kuvaus $f:A \to \R$ jatkuva pisteessä $x \in A$, jos jokaiselle annetulle $\epsilon > 0$ on olemassa $\delta > 0$ siten, että $|f(x) - f(y)| < \epsilon$ aina kun $y \in A$ ja $\|x - y \| < \delta$ eli
\[
\lim_{y \to x}f(y) = f(x)
\]

Olkoon $A \subset \R^n$ ja olkoon $f:A \to \R$ kuvaus. Kuvaus $f$ on jatkuva pisteessä $a \in A$, jos ja vain jos
\[
f(x_k) \to f(a)
\]
kaikilla jonoilla $(x_k)$, jolla $x_k \to a$, $x_k \in A$.

\begin{example}
Onko kuvaus $f: \R^2 \to \R$ jatkuva?
\[
f(x_1, x_2 = \begin{cases}
\frac{x_1 x_2^2}{x_1^2 + x_2^2} & (x_1, x_2) \neq (0,0) \\
0 & (x_1, x_2) = (0,0) \\
\end{cases}
\]

$x_1 = kx_2^2$, $f(x_1, x_2) = f(kx_2^2, x_2) = \frac{kx_2^4}{k^2x_2 4 + x_2 4} = \frac{k}{1 + k^2} \neq 0$

Ei ole jatkuva origossa!
\end{example}

\section{Reaaliarvoisten kuvausten korkeammat osittaisderivaatat ja lokaalit ääriarvot}
\subsection{Korkeammat osittaisderivaatat}
Reaaliarvoisella funktiolla $f:G \to \R$, $G \subset \R^n$ avoin, olkoon pisteen $x_0 \in G$ avoimessa ympäristössä $U \subset G$ osittaisderivaatta $\partial_i f(x)$, $\forall x \in U$. Jos funktiolle $\partial_i f: U \to \R$ on pisteessä $x_0$ osittaisderivaatta $\partial_j(\partial_j f)(x_0)$, niin tätä sanotaan funktion $f$ toisen kertaluvun osittaisderivaataksi pisteessä $x_0$ ja merkitään
\[
\partial_j (\partial_i f)(x_0) = \partial_j \partial_i f(x_0) = \frac{\partial^2 f}{\partial x_j \partial x_i}(x_0) = D_{ji}f(x_0) = \partial_{ji}f(x_0)
\]
\begin{example}
Funktiolle $f: \R^3 \to \R$, $f(x) = x_1 \sin x_3 + x_2^2$

$\partial_1 f(x) = \sin x_3$, $\partial_1 \partial_2 f(x) = 0$
\end{example}

Funktion $f:G \to \R$, $G \subset \R^n$ avoin, on $k$ kertaa (osittais)derivoituva pisteessä $x_0 \in G$, jos sen kaikki osittaisderivaatat kertalukua $\leq k$ ovat olemassa pisteessä $x_0$.

Funktio $f: G \to \R$, $G \subset \R^n$ avoin, on $k$ kertaa jatkuvasti derivoituva avoimessa joukossa $G$ eli $f$ on $C^k$-funktio, merkitään $f \in C^k(G)$, jos sen kaikki osittaisderivaatat kertalukua $\leq k$ ovat olemassa ja jatkuvia avoimessa joukossa $G$.

Funktion $f: G \to \R$, $G \subset \R^n$ avoin, on äärettömän monta kertaa jatkuvasti derivoituva avoimessa joukossa $G$ eli $f$ on $C^{\infty}$-funktio, merkitään $f \in C^{\infty}(G)$, jos sen kaikkien kertalukujen osittaisderivaatat ovat olemassa ja jatkuvia avoimessa joukossa $G$.

Lause: Jos $C^1$-funktiolle $f:G \to \R, G \subset \R^n$ avoin, on psiteen $x_0 \in G$ avoimessa ympäristössä jatkuva osittaisderivaatta $\partial_j \partial_i f$, niin silloin myös $\partial_i \partial_j f(x_0)$ on olemassa ja $\partial_i \partial_j f(x_0) = \partial_j \partial_i f(x_0)$.

Korollaari: Jos $f:G \to \R$, $G \subset \R^n$ avoin, on $C^k$-funktio ja $\partial_i_1 \partial_i_2 \dots \partial_i_k f$, $1 \leq ij \leq n$, $j = 1, \dots, k$, on jokin $k$. kertaluvun osittaisderivaatta, niin 
\[
\partial_i_1 \partial_i_2 \dots \partial_i_k f = \partial_r_1 \partial_r_2 \dots \partial_r_k f
\]

\begin{enumerate}

\end{enumerate}

\int_B f = \int_{g(A)}f = \int_A f(g(y))|J_g(y)|dy
\]

Olkoon $g: W \to \R^n$ jatkuvasti differentioituva, $W \subset \R^n$, avoin. Oletetaan, että $U \subset W$ on rajoitettu, $\overline{U} \subset W$, $\partial U$ on n-nollajoukko ja $g:(\textrm{int}U)\setminus N \to g((\textrm{int}U) \setminus N)$ on diffeomorfismi, missä $N \subset \overline{U}$ on n-nollajoukko. 

Jos rajoitettu kuvaus $f: g(U) \to \R$ on integroituva, niin $(f \circ g)|J_g|: U \to \R$ on integroituva ja $\int_{g(U)}f = \int_U(f \circ g)|J_g|$ eli $\int_{g(U)}f(x)dx = \int_U (f \circ g)(u)|J_g(w)|du$

\subsection{Sylinterikoordinaattikuvaus avaruudessa $\R^3$}

J_g(r, \theta, z) = \begin{vmatrix}\cos \theta & -r \sin \theta & 0 \\\sin \theta & r \cos \theta & 0 \\ 0 & 0 & 1\end{vmatrix} = r

ja $g(W) = \R^3 \setminus \{(x, 0, z) \in \R^3: x \geq 0\}$

\section{Epäoleellinen integraali}
\subsection{Integraali yli avaruuden $\R^n$, $n \geq 2$}
Olkoon $N > 0$ kiinnitetty. Tarkastellaan origokeskistä kuutiota $K_n = \{(x_1, \dots, x_n) \in \R^n|(x_j) \leq N/2, j=1, \dots, n \}$, joiden sivun pituus on $N$. Olkoon $f: \R^n \to \R_+$ jatkuva funktio. Jos on olemassa raja-arvo $\lim_{N \to \infty}\int_{K_N}f(x)dx$, niin $\int_{\R^n}f(x)dx = \lim_{N \to \infty}\int_{K_N}f(x)dx$

Lause: Iteroitu integraali. Jos $f$ on jatkuva tasossa $\R^2$ ja $|f(x)| \leq \frac{A}{1 + |x|^3}, A > 0$ vakio, kaikilla $x$, niin
\[
\int_{\R^2}f(x)dx = \int_{\R}\int_{\R}f(x_1, x_2)dx_2 dx_1
\]
Jos suppeneminen tiedetään, niin $\int_D f = \int_D f^+ - \int_D f^-$

\section{Käyräintegraalit}
\subsection{Reaaliarvoisen funktion käyräintegraali}
Määritelmä: Olkoon $J = [a, b] \subset \R$ ja $A \subset \R^n$ ja $\gamma:[a, b] \to A$ jatkuvasti differentioituva polku. Jos $f: A \to \R$ on Riemann-integroituva, niin funktion $f$ käyräintegraali polun $\gamma$ kaaren pituuden suhteen eli reaaliarvoisen funktion $f$ käyräintegraali polun $\gamma$ suhteen (eli polkuintegraali) on:
\[
\int_{\gamma}fds = \int_J (f \circ \gamma)\|\gamma'\| = \int_a^b f(\gamma(t))\|\gamma'(t)\|dt
\]

\subsection{Vektorikentän käyräintegraali}
Määritelmä: Vektorikenttä. Olkoon $D \subset \R^n$. Vektorikenttä avaruudessa $\R^n$ on funktio $F$, joka liittää jokaiseen pisteeseen $x \in D$ vektorin $F(x) \in \R^n$. 

Vektorikenttä $F$ (oletetaan, että $F$ on jatkuva, eli $F \in C$) on konservatiivinen, jos se on jonkin reaaliarvoisen vektorifunktion $u \in C^1$ gradientti, eli on olemassa $U$ siten, että $\nabla u = F$. Funktio $u$ on vektorikentän $F$ potentiaali. 

Määritelmä: Vektorikentän käyräintegraali. Olkoon $\gamma = (\gamma_1, \dots, \gamma_n): [a, b] \to \R^n$ $C^1$-polku. Olkoon $f=(f_1, \dots, f_n): \R^n \to \R^n$ vektorikenttä ja $f$ jatkuva. Vektorikentän käyräintegraali polun $\gamma$ suhteen
\[
\int_{\gamma}f \cdot d\overline{s} = \int_J(f \circ g|\gamma') = \int_a^b f(\gamma(t))\cdot\gamma'(t) dt
\]

\begin{example}
$f(x, y) = xy$, $y=x$ pitkin $(0,0) \to (1,1)$

$\gamma(t) = (t, t)$, $[0,1]$, $t=0 \to t=1$

$\int_{\gamma}fds = \int_0^1 t^2 \|(1,1)\|dt = \int_0^1 t^2 \sqrt{1^2 + 1^2}dt$

$\sqrt{2}\int_0^1 t^2 dt = \sqrt{2}\bigg|_0^1 1/2 t^3 = \sqrt{2}/3$
\end{example}
