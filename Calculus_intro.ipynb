\section{Invers funktion}
\subsection{Definition}
Funktionerna $f:A \to B$ och $g: B \to A$ är varandras inversa funktioner om det för alla $x \in A$ gäller att $g(f(x)) = x$ o ch för aööa $y \in B$ gäller att $f(g(y)) = y$. Funktionerna $f: A \to B$ och $g: B \to A$ är alltså varandras inversa funktioner när
\[
f \circ g = g \circ f
\]
\[
f(x) \circ f^{-1}(x) = f^{-1}(x) \circ f(x)
\]

\begin{example}
$f(x) = x^2 + 1$, $y = x^2 + 1$, $y -1 = x^2$, $x = \sqrt{y -1}$, Variabelbyte $y = \sqrt{x - 1}$

$f^{-1}(x) = \sqrt{x - 1}$
\end{example}

\subsection{Egenskaper}
Spegling i $y = x$, Kontinuitet och monotonitet bibehålls, $D_f = V_f^{-1}$ och $V_f = D_f^{-1}$

Om en funktion är entydigt monoton har den en invers funktion!

\subsection{Derivata}
\begin{proof}
Vi vet att $y - a = (f^{-1})'(b)(x-b)$ och $y -b = f'(a)(x-a)$. 
Funktionen: 
\[y - b O f'(a)(x - a)
\]
\[
x - b = f'(a)(y -a)
\]
\[
\frac{1}{f'(a)}(x - b) = y - a
\]
Vi vet också
\[
y - a = (f^{-1})'(b)(x - b)
\]
Alltså 
\[
\frac{1}{f'(a)} = (f^{-1})'(b)
\]
\[
(f^{-1})'(y_0) = \frac{1}{f'(x_0)}
\]
\end{proof}

\begin{example}
$(f^{-1})'(8)$, $f:[1,5] \tp \R$, $f =\frac{12x}{x + 1}$

$8 = y_0 = f(x_0)$, $8 = 12x_0/(x_0 + 1)$, $x_0 = 2$

$(f^{-1})'(8) = \frac{1}{f'(2)} = 3/4$
\end{example}

\section{Kontinuerliga fördelningar}
\subsection{Normalfördelning}
$P(X \leq a) = \phi(a)$

$P(X \geq a) = 1 - \phi(a)$

$P(X \leq -a) = 1 - \phi(a)$

\subsection{Probability distribution function}
Diskret stokastisk variabel = frekvensfunktion

$f(x) \geq 0$, $\sum_{\forall i} f(x_i) = 1$

Kontinuerlig stokastisk variabel = täthetsfunktion

$f(x) \geq 0$ och $\int_{-\infty}^{\infty}f(x)dx = 1$

\subsection{Cumulative distribution function}
Diskret = fördelningsfunktion

$P(X \leq t) = \sum_{i \leq t}f(x_i) = F(t)$

Kontinuerlig = fördelningsfunktion

$P(X \leq t) = \int_{-\infty}^t f(x) dx = F(t)$

$P(a \leq X \leq b) = \int_a^b f(x) dx =F(b) - F(a)$

\subsection{Lägesmått}
Diskret: väntevärde $\mu = \sum_{i=1}^n p_i x_i$, varians $\sigma^2 = \sum_{i=1}^n p_i(x_i - \mu)^2$

Kontinuerlig: $\mu = \int_{-\infty}^{\infty}x f(x) dx$, $\sigma^2 = \int_{-\infty}^{\infty}f(x)(x - \mu)^2dx$

\section{Talföljder och serier}
\subsection{Gränsvärde för en talföljd}
\[
\lim_{n \to \infty} = L \iff \forall \epsilon > 0 \exists N \in Z_+: |a_n - L| < \epsilon, n \geq \N
\]

\subsection{Konvergens för en geometrisk talföljd}
$a, aq, aq^2, \dots$
\begin{enumerate}
    \item $-1 < q \leq 1$
    \item $a_1 = 0$
\end{enumerate}

\subsection{Serie = summan i en oändlig talföljd}
\subsubsection{Geometrisk serie}
\[
\sum_{k=1}^{\infty}a_1q^{k-1} = a_1 + a_1q + a_1q^2 + \dots = \frac{a_1(1 - q^n)}{1-q}
\]
\[
\lim_{n \to \infty}S_n = \lim_{n \to \infty} \frac{a_1(1-q^n)}{1-q} = \frac{a_1}{1-q}
\]
Kongruens:
\begin{enumerate}
    \item $-1 < q < 1$
    \item $a_1 = 0$
\end{enumerate}

\subsusbection{Serier}
Divergenstest
\begin{enumerate}
    \item $\lim_{n \to \infty}a_n \neq 0$ eller saknas $\to$ div.
    \item $\lim a_n = 0$ $\to$ $\lim_{n \to \infty}S_n$ delsumma, saknas $\to$ divergerar, $S$ existerar och serien konvergerar
\end{enumerate}
