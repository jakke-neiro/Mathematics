\subsection{Mononuclear reaction}
$A \xrightarrow{\mu} B$ or $A \xrightarrow{\mu} B + C$ 
where $\mu > 0$ is the probability per unit of time (i.e. reaction rate) that A undergoes a reaction ($\mu$ is a rate with dimension 1/time. 

Let $T \geq 0$ be the time at which a reaction occurs, and define $P(t) = \textrm{Prob}(T > t)$, which is the same as the probability that no reaction has happened at time $t$ yet. 

$P(t + h) = P(t) - \mu h P(t)$

$\to \frac{P(t+h) - P(t)}{h} = -\mu P(t)$

$\to \frac{dP(t)}{dt} = -\mu P(t) \to P(t) = e^{\mu t}, P(0) = 1$

The time $T$ till a reaction has cumulative distribution function
\[
F(t) = \textrm{Prob}(T \leq t) = 1 - e^{-\mu t}
\]
and probability density
\[
f(t) = F'(t) = \mu e^{-\mu t}
\]
In other words, $T$ has an exponential distribution with rate $\mu$, and hence mean waiting time is $\mu^{-1}$.

\subsection{Mononuclear reaction with time-variable rate}
$A \xrightarrow{\mu(t)} B$ etc.
\[
\frac{dP(t)}{dt} = - \mu(t)P(t) \to P(t) = e^{- \int_0^t \mu(z)dz}
\]
\[
\to F(t) = 1 - e^{-\int_0^t \mu(z) dz}
\]
\[
\to f(t) = \mu(t)e^{-\int_0^t \mu(z) dz}
\]
Not exponential distribution

\subsection{Network of monomolecular reactions}
$A$ is involved in several independent mononuclear reactions, e.g. $A \xrightarrow{\mu} B$ and $A \xrightarrow{\nu} C + D$. The total reaction rate is $\mu + \nu$ and hence
\[
\frac{d P(t)}{dt} = - (\mu + \nu)P(t) = -\mu P(t) - \nu P(t)
\]
i.e. we can simply add up the ffects of the separate reactions. 
\[
\to F(t) = 1 - e^{-(\mu + \nu)t}
\]
\[
\to f(t) = (\mu + \nu)e^{-(\mu + \nu)t}
\]
\subsection{From the individual level to the population level}
The strong law of the large numbers allows us in large populations to equate probabilities with proportions, i.e. $P(t) = A(t)/A(0)$, where $A(t)$ is the population density of $A$ particles. 
\[
P(t) = \frac{A(t)}{A(0)}
\]
\[
A(t) = A(0)P(t)
\]
\[
\frac{d A(t)}{dt} = A(0) \frac{dP(t)}{dt} = -\mu A(0) P(t) = -\mu A(t)
\]
and since $B$ increases at the same rate that $A$ decreases
\[
\frac{dB(t)}{dt} = \mu A(t)
\]\begin{example}
$A \xrightarrow{\mu} B + C$

$\frac{dA}{dt} = -\mu A, \frac{dB}{dt} = \mu A, \frac{dC}{dt} = \mu A$
\end{example}

\begin{example}
$A \xrightarrow{\mu} A + B$, $B \xrightarrow{\nu} A$

$\frac{dA}{dt} = \nu B$, $\frac{dB}{dt} = \mu A - \nu B$
\end{example}

\subsection{The bimolecular reaction}
Modelling interactions $A + B \xrightarrow{\alpha} C$

Empirical law of mass-action: in a bimolecular reaction, the probability per unit of time of one reactant undergoing a reaction is proportional to the concentration (or population density) of the other reactant.

$P_a(t)$ : prob that A has not yet undergone a reaction

By the law of mass-action:

$P_A(t+h) = P_A(t) - \alpha B(t)hP_A(t)$

$P_A(0) = 0$ 

$\frac{P_A(t+h) - P_A(t)}{h} = -\alpha B(t)P_A(t) $

$ \to \frac{dP_A(t)}{dt} = -\alpha B(t)P_A(t), P_A(0) = 1$

Likewise for $P_B(t)$:

$\frac{dP_B(t)}{dt} = - \alpha A(t) P_B(t), P_B(0) = 1$

This is a model on the individual level. We use the law of large numbers to get the model
\[
\frac{dA(t)}{dt} = - \alpha A(t)B(t) 
\]
\[
\frac{dB(t)}{dt} = - \alpha A(t)B(t)
\]

\begin{example}
i-states: $X$ free prey, $S$ searching predator, $H$ handing predator

i-processes: $X + S \xrightarrow{\alpha} H$ (prey capture), $H \xrightarrow{\beta} X + S$ (prey escape), $H \xrightarrow{\gamma} S$ (prey killed)

p-states: $X$ pop. density of $X$, $S$ pop. density of $S$, $H$ pop. density of $H$

Population-level model

\frac{dS}{dt} = - \alpha XS + \beta H + \gamma H

\begin{example}
Logistic equation

i-state: $X$ alive individual

\end{example}

\begin{example}
$E$ empty site/territory, $X$ occupied site/territory, $Y$ individual without site/territory

$X \xrighytarrow{\alpha} X + Y$, $Y + E \xrightarrow{\beta} X$ (colonization), $X \xrightarrow{\gamma} E$ (death of $X$), $Y \xrightarrow{\delta}$ dead (death of $Y$)

p-level: $E, X, Y$ pop. densities of $E, X, Y$
\[
\frac{dE}{dt} = -\beta EY + \gamma X
\]
\[
\frac{dX}{dt} = \beta EY - \gamma X
\]
\[
\frac{dY}{dt} = \alpha X - \beta EY - \delta Y
\]
\end{example}

Conservation relation: note that $\frac{dX}{dt} + \frac{dE}{dt}$, which means that $X + E = E_0$ is constant. 

Then we get
\[
\begin{cases}
\frac{dX}{dt} = \beta (E_0 - X)\gamma - \gamma X \\
\frac{dY}{dt} = \alpha X - \beta(E_0 - X)Y - \delta Y
\end{cases}
\]

Conservation relations are used to reduce the number of equations in an ODE system.

\subsection{Timescale separation}
Some $i$-level processes may be significantly faster than other. If the differences are really big, we may solve the pop. model involving only fast processes first and add the slow processes later.

This is also a standard method to reduce the number of equations

For example, we can suppose that reproduction and death outside sites are fast processes

This can be expressed by dividing corresponding rares by a small $\epsilon > 0$. 

$X \xrightarrow{\alpha/\epsilon} X + Y$ (fast)

$Y + E \xrightarrow{\beta} X$ (slow)

$X \xrightarrow{\gamma} E$ (slow)

$Y \xrightarrow{\delta/\epsilon}$ dead (fast)

The system is then
\[
\begin{cases}
\frac{dX}{dt} = \beta (E_0 - X)\gamma - \gamma X \\
\frac{dY}{dt} = \frac{\alpha}{\epsilon} X - \beta(E_0 - X)Y - \frac{\delta}{\epsilon} Y
\end{cases}
\]

$Y$ is a fast variable and $X$ is a slow variable

Multiplication of the equation by $\epsilon$ gives the system

\[
\begin{cases}
\frac{dX}{dt} = \beta (E_0 - X)\gamma - \gamma X \\
\epsilon \frac{dY}{dt} = \alpha X - \epsilon \beta(E_0 - X)Y - \delta Y
\end{cases}
\]

Letting $\epsilon \to 0$, we get the following differentio-algebraic system
\[
\begin{cases}
\frac{dX}{dt} = \beta(E_0 - X)Y - \gamma X \\
0 = \alpha X - \delta Y
\end{cases}
\]

We first solve the second equation for $Y$, which gives 
\[
y = \frac{\alpha}{\beta}X
\]
and we substitute that solution into the equation for $\frac{dX}{dt}$
\[
\frac{dX}{dt} = \frac{\alpha \beta}{\delta}(E_0 - X)X - YX
\]
which we can also write as
\[
\frac{dX}{dt} = rx(1 - \frac{X}{K})
\]
\[
r = \frac{\alpha \beta}{\delta}E_0 - \gamma
\]
\[
k = E_0 - \frac{\gamma \delta}{\alpha \beta}
\]
which is the logistic equation. Note that this is a different derivation of the logistic involving different $i$-level processes. The interpretation of the model parameters is deiiferent!

\subsection{A more careul look at timescale separation}
\[
\begin{cases}
\frac{dX}{dt} = \beta (E_0 - X)\gamma - \gamma X \\
\frac{dY}{dt} = \frac{\alpha}{\epsilon} X - \beta(E_0 - X)Y - \frac{\delta}{\epsilon} Y
\end{cases}
\]

Here $\gamma$ is the fast variable. We introduce the fast time variable $t^* = t/\epsilon$ and rewrite the system for $t^*$ using that
\[
\frac{d}{dt^*} = \frac{d}{dt}\frac{dt}{dt^*} = \epsilon \frac{d}{dt}
\]
which gives
\[
\begin{cases}
\frac{dX}{dt^*} = \epsilon \beta (E_0 - X)Y - \epsilon \gamma X \\
\frac{dY}{dt^*} = \alpha x - \epsilon \beta (E_0 - X)Y - \delta Y
\end{cases}
\]
Letting $\epsilon \to =$, we get
\[
\begin{cases}
\frac{dX}{dt^*} = 0 \\
\frac{dY}{dt^*} = \alpha X - \delta Y
\end{cases}
\]
Since fast time $t^*$ became infinitely fast, we can still see how $Y$ changes. Thus, $Y$ converges to $\frac{\alpha X}{\delta}$ in fast time, which justifies the substitution $Y = \frac{\alpha X}{\delta}$

Singular perturbation theory: infinitely fast system retains the topological properties of the system without complete separation, provided the equilibrium in $t^*$-time for $\epsilon \to 0$ is hyperbolically stable.

Linearization of the system at the interior equilibrium $(\overline{x}, \overline{y})$ gives
\[
\frac{d}{dt}\begin{bmatrix}x - \overline{x} \\ y - \overline{y} \end{bmatrix} = 
\begin{bmatrix}
-\frac{r \overline{x}}{K} + r(1 - \frac{\overline{X}}{K}) - \beta \overline{Y} & -\beta \overline{X} \\
p \beta \overline{Y} & p \beta \overline{X} - \delta \\
\end{bmatrix}
\begin{bmatrix}
X - \overline{X} \\
Y - \overline{Y}
\end{bmatrix}
\]

At the equilibrium, $\frac{dx}{dt} = \frac{dy}{dt} = 0$, and therefore $r(1 - \frac{\overline{x}}{K}) - \beta\overline{Y} = 0$ and $p \beta \overline{X} - \delta = 0$

The Jacobian matrix thus simplifies into
\[
\begin{bmatrix}
-\frac{r \overline{X}}{K} & - \beta \overline{X} \\
p \beta \overline{Y} & 0 \\
\end{bmatrix}
\]
with $\textrm{det}J = p \beta^2 \overline{X}\overline{Y} > 0$ and $\textrm{tr}(J) = -\frac{r \overline{x}}{K} < 0 $

$\to (\overline{x}, \overline{y})$ is locally attracting

\section{Appendix A: Local stability analysis}
Consider the system in $\R^n$ ($n \geq 1$)
\[
\frac{dy}{dt} = f(y)
\]
where $f: \R^n \to \R^n$ is conti. diff. Suppose that $\hat{y}$ is an equilibrium for the equation, i.e. $f(\hat{y}) = 0$.

It can be proven that the equation is locally topologically equivalent to the system
\[
\frac{dx}{dt} = Df(\hat{y})x
\]
whenever the eigenvalues of the Jacobian matrix $Df(\hat{y})$ have non-zero real parts. 

\subsection{Linear systems}
Consider the linear system
\[
\frac{dx}{dt} = Ax, x(0) = x_0
\]
where $x \in \R^n (n \geq 1)$, $A \in \R^{n \times n}$. The formal solution is
\[
x(t) = e^{At}x_0, t \geq 0
\]
where
\[
e^{At} = \sum_{k=0}^{\infty}\frac{t^k}{k!}A^k
\]
which is known to converge for all $A \in \R^{n \times n}$

But what do the orbits actually look like? Under what conditions do the orbits converge to zero, the only equilibrium?

Definition: a matrix $A \in \R^{n \times n}$ is diagonilizable if there exist matrices $B, \Lambda \in \C^{n \times n}$, such that $det B \neq 0$, $\Lambda$ is diagonal, and $A = B \Lambda B^{-1}$.

Proposition: suppose that a matrix $A \in \R^{n \times n}$ has $n$ distinct eigenvalues $\lambda_1, \dots, \lambda_n \in \C$. Then the corresponding eigenvectors $b_1, \dots, b_n \in \C^n$ are linearly independent, and
\[
A = B \Lambda B^{-1}
\]
where $\Lambda = +textrm{diag}(\lambda_1, \dots, \lambda_2)$ and $B = (b_1, \dots, b_n)$.

Remark: the proposition not only gives a condition such that a square matrix is diagonalizable, but it also shows that being diagonalizable is the generic situation.

Proof of proposition: Let $\lambda_1, \dots, \lambda_n \in \C$ be distinct eigenvalues of $A$ with corresponding eigenvectors $b_1, \dots, b_n \in \C^n$
\[
Ab_i = \lambda_i b_i \forall i
\]
Define $B = (b_1, \dots, b_n)$, then
\[
AB = B\Lambda
\]
If we can show that $b_1, \dots, b_n$ are independent, then it follows immediately that $A 0 B \Lambda B^{-1}$. 

To reach a contradiction, suppose $b_1, \dots, b_n$ are not independent. Then there exists an integer $m < n$ such that $b_1, \dots, b_m$ are independent while $b_{1}, \dots, b_m, b_{m+1}$ are not. 

Then there exists scalars $\alpha_1, \dots, \alpha_m \in \C$ (not all zeros) such that
\[
b_{m+1} = \alpha_1b_1 + \dots + \alpha_m b_m
\]
Left multiplication with $A$ gives
\[
\lambda_{m+1}b_{m+1} = \alpha_1 \lambda_1 b_1 + \dots + \alpha_m \lambda_m b_m
\]
Substitution of the first equation into the lefthand-side of the second equation and some rearrangements of terms gives
\[
\alpha_1(\lambda_{m+1} - \lambda_1)b_1 + \dots + \alpha_m(\lambda_{m+1} - \lambda_m)b_m = 0
\]
Since $b_1, \dots, b_m$ are assumed to  be independent, it follows that $\alpha_i(\lambda_{m+1} - \lambda_i)=0$ for $i = 1, \dots, m$., Since not all $\alpha_i$ are zero, there exists an $i_0 \in \{1, \dots, m\}$ such that $\alpha_{i_0} \neq 0$, and hence $\lambda_{m+1} - \lambda_{i_0} = 0$. But that contradicts the assumption that $\lambda_1, \dots, \lambda_n$ are all distinct. 

Proposition: Suppose $A \in \R^{n \times n}$ has $n$ distinct eigenvalues $\lambda_1, \dots, \lambda_n \in \C$ and corresponding eigenvalues $b_1, \dots, b_n \in \C^n$. Then the solution of
\[
\frac{dx}{dt} = Ax, x(0) = x_0
\]
can be written as
\[
x(t) = \beta_1e^{\lambda_1 t}b_1 + \dots + \beta_n e^{\lambda_n t}b_n
\]
for any given $\beta_1, \dots, \beta_n \in \C$. 

Proof: Let $\beta_1, \dots, \beta_n \in \C$ be such that $x_0 = \beta_1 b_1 + \dots \beta_n b_n$, which is possible, because by the previous proposition $b_1, \dots b_n$ form a basis of $\C^n$. Let $x(t)$ be as in the last equation, Then $x\cos = x_0$, and
\[
\frac{d x(t)}{dt} = \lambda_1 \beta_1 e^{\lambda_1 t}b_1 + \dots + \lambda_n \beta_n e^{\lambda_n t}b_n = 
\]
\[
= \beta_1 e^{\lambda_1 t}Ab_1 + \dots + \beta_n e^{\lambda_n t}A b_n = 
\]
\[
=
A (\beta_1e^{\lambda_1 t}b_1 + \dots + \beta_n e^{\lambda_n t}b_n) = 
\]
\[
= A x(t)
\]

Corollary: under the conditions of the previous proposition: $Re(\lambda_i) < 0 \forall i \to x(t) \to 0$ as $t \to \infty$. $\exists i \in \{1, \dots, n \}: Re(\lambda_i) > 0 \to \|x(t)\| \to \infty$ as $t \to \infty$. $\beta_1 \neq 0, Re(\lambda_1) > Re(\lambda_i) \forall i \neq 1 \to e^{-\lambda_1 t}x(t) \to \beta_1 b_1$ as $t \to \infty \infty$. Eigenvectors are invariant under $\frac{dx}{dt} = Ax$. 

\subsection{The case $n=2$}
$\lambda_1$ and $\lambda_2$ real and distinct: $\lambda_2 < \lambda_1 < 0$ stable node, $0 < \lambda_2 < \lambda_1$ unstable node, $\lambda_2 < 0 < \lambda_1$ saddle

$\lambda_1$ and $\lambda_2$ complex and distinct:

$Ab_1 = \lambda_1 b_1 \to A \overline{b}_1 = \overline{\lambda}_1 \overline{b}_1$ $(A \in \R^{2 \times 2})$.

Thus, $\lambda_1$ and $\lambda_2$ are complex conjugates, and so are the corresponding eigenvalues. What about the coefficients $\beta_1$ and $\beta_2$?

$\lambda_1 = p + iq$, $\lambda_2 = p - iq$, $p, q \in \R$

$b_1 = \begin{bmatrix}ue^{i \phi} \\ ve^{i \pis}\end{bmatrix}$, $b_2 = \begin{bmatrix}ue^{-i \phi} \\ ve^{-i \psi}\end{bmatrix}, u, v, \phi, \psi \in \R$. 

$x = \begin{bmatrix}x_1 \\ x_2 \end{bmatrix}, x_1, x_2 \in \R$

\[
x_1(t) = ue^{pt}[\beta_1 e^{i(qt + \phi)} + \beta_2 e^{-i(qt + \phi)}]
\]
$x_1(t) \in \R$, $ue^{pt} \in \R$, last term must also be real $\forall t$. 

Let $t > 0$ be such that $qt + \phi = 2k\pi$ for some $k \in \N$ (possible because $\lambda_1 \neq \lambda_2, \lambda_1 = \overline{\lambda_1} \to q \neq 0)$

$\beta_1 + \beta_2 \in \R \to Im \beta_1 = - Im \beta_2$

Let $t > 0$ be such that $qt + \phi = \frac{1}{2}\pi + 2k\pi$ for some $k \in \N$. $i \beta_1 - i \beta_2 \in \R \to Re \beta_1 = Re \beta_2$. Conclusion $\beta_1 = \overline{b}_2$.

\subsection{Local stability analysis}
Consider the system
\[
\frac{dX}{dt} = f(x)
\]
where $f: \R^n \to \R^n$, $n \geq 1$. Let $\overline{x} \in \R^n$ be such that $0 = f(\overline{x})$, i.e. $\overline{x}$ is an equilibium of the differential system. 

Linearization of the differential system around $\overline{x}$ gives
\[
\frac{dX}{dt} = f'(X)(X - \overline{X})
\]
where $f'(\overline{X})$ is the matrix of partial derivatives of $f$ at $\overline{X}$ (also called the Jacobian matrix of $f$). Note that $\overline{X}$ is also an equilibrium of the llnearized system.

The Hartman-Grobman theorem implies that if all eigenvalues of $f'(\overline{x})$ have a non-zero real part, then stability of $\overline{x}$ in the linearized system implies stability of $\overline{x}$ in the non-linearized system and instability of $\overline{X}$ in the linearized system implies instability of $\overline{X}$ in the non-linearized system.

(The Harman-Grobman theory says a bit more but we leave thet for later)

The linearized system is equivalent to 
\[
\frac{dz}{dt} = Az
\]
with $z = X - \overline{X}$ and $A = f'(\overline{X})$. We study the stability of the equilibrium $\overline{z} = 0$ in the linear system. 

Let $u_1, \dots, u_n \in \C^n$ be eigenvectors of $A$ with corresponding eigenvalues $\lambda_1, \dots, \lambda_n$, and define the matrices
\[
U = \begin{bmatrix}u_1 & \dots & u_n \end{bmatrix} \in \C^{n \times n}
\]
\[
\Lambda = \textrm{diag}\begin{bmatrix}\lambda_1, \dots, \lambda_n \end{bmatrix} \in \C^{n \times n}
\]
Thus we can write
\[
AU = U\Lambda
\]
Suppose $U$ is non-singular. Then right multiplication with $U^{-1}$ gives
\[
A = U \Lambda U^{-1}
\]
(A sufficient condition for $U$ to be non-singular os that all eigenvalues of $A$ are distinct).

Thus we get
\[
\frac{dz}{dt} = U \Lambda U^{-1}z
\]
Left-multiplication with $U^{-1}$ on both sides gives
\[
\frac{dV}{dt} = \Lambda v
\]
where $v = U^{-1}z$.

Since $\Lambda$ is a diagonal matrix, the system is jst a collection of $n$ uncoupled linear ODEs. So, the solution is trivial
\[
v(t) = \begin{bmatrix}e^{\lambda_1 t} & \dots & 0 \\
\vdots & \ddots & \vdots \\
0 & \dots & e^{\lambda_n t} \end{bmatrix}
\]
or equivalently, $v_i(t) = v_i(0)e^{\lambda_i t} (i = 1, \dotsm n)$ where $v = \begin{bmatrix}v_1 & \dots & v_n \end{bmatrix} \in \C^n$, Taking absolute values on both sides, we get
\[
|v_i(t)| = |v_i(0)|e^{Re \lambda_i t}
\]
Conclusion: 

If $Re \lambda_i , 0$ for all $i$, then $|v_i(t)| \to 0$ as $t \to \infty$ for all $i$, and hence $v(t) \to 0$ as $t \to \infty$, and so $z(t) = Uv(t) \to 0$ and $x(t) = \overline{x} + z(t) \to \overline{x}$.

If $Re \lambda_{i_0} > 0$ for at least one $i_0$, $|v_{i_0}(t)| \to \infty$ as $t \to \infty$ and so $\| v(t) \| \to \infty$ and $\infty$ and so $\|x(t) - \overline{x} \| = \|z(t)\| = \|U v(t) \| \to \infty$. Of course, as $\| x(t) \to \overline{x} \| \to \infty$, the linearization is not accurate, but the point is that $x(t)$ does not stay close to $\overline{x}$ as $t \to \infty$.

Apply Hartman-Grobman: suppose $\overline{x}$ is a hyperbolic equilibrium of the nonlinearized system (i. e. all eigenvalues of $f'(x)$ have a non-zero real part). 

Then: $\overline{x}$ is attracting if all eigenvalues of $f'(x)$ have a negative real part. $\overline{x}$ is unstable if at least one eigenvalue of $f'(x)$ has a positive real part.
