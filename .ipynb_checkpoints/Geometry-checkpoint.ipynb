\section{Plangeometriska grundbegrepp}
\subsection{Punkt och linje}
Punkten A, linjen AB, linjen l

Två linjer: en punkt gemensam $to$ skär

Två punkter gemensamma $\to$ sammanfaller

Två linjer är parallella när de inte skär varandra; $m \parallel l$

Två linjer är varandras normaler om de står vinkelrätt mot varandra $m \perp l$

Stråle $l$, sträcka $a$

Polygon

Mittpunktsnormalen till en sträcka är den normal som går genom mittpunkten på sträckan.

Varje punkt på mittpunktsnormalen till en sträcka är lika långt från vardera ändpunkten av sträckan. Varje punkt i planet som är lika långt från vardera ändpunkten på en sträcka ligger på sträckans mittpunktsnormal

\subsection{Delning av en sträcka}
$P$ delar träckan i föhållandet $5:3$ invändigt (totala 8)

$P$ delar sträckans förlängning i förhållandet $2:7$

\subsection{Vinkel}
Två strålar som utgår från samma punkt, vinkeln $\alpha$, $\spericalangle ABC$, $\sphericalangle B$

Vänster vinkelben, höger vinkelben

Full vinkel $360^o$

\begin{enumerate}
    \item Nollvinkel: $0^o$
    \item Spetsig vinkel: $0^o < \alpha < 90^o$
    \item Rät vinkel: $90^o$
    \item Trubbig vinkel: $90^0 < \alpha < 180^o$
    \item Rak vinkel: $180^o$
    \item Konvex vinkel: $180^o < \alpha < 360^o$
    \item Full vinkel: $360^o$
    \item Komplementvinklar: $\alpha + \beta = 90^o$
    \item Supplementvinklar: $\alpha + \beta = 180^o$
    \item Explementvinkalr: $\alpha + \beta = 360^o$
\end{enumerate}

$\alpha + \beta = 180^o$, gemensam vinkelspets och ett gemensamt vinkelben = sidovinklar

Bisektris: en stråle från vinkelspetsen som delar vinkeln i två lika stora delar

Vertikalvinklar

$\alpha = \beta$ Likbelägna vinklar omm $l \parallel m$

$\alpha = \gamma$ Alternatvinklar om $l \parallel m$

\section{Månghörning}
\subsection{Grundbegrepp}
Polygon: sammanfogade sträckor

Månghörning: en planfigur begränsad av en sluten polygon som inte skär sig själv

Omskriven cirkel: tnagerar alla hörn

Inskriven cirkel: tangerar alla sidor

Vinkelsumman: $(n-2)*180^o$

Antal diagonaler: $\frac{n(n-3)}{2}$

\subsection{Triangel}
$c$ långa sidan, $\to$ $\gamma$ största vinkeln

Vinkelsumman $180^o = \alpha + \beta + \gamma$

Triangelns area: $\frac{1}{2}bh$

Likbent triangel: två sidor lika långa

Liksidig triangel: alla sidor lika långa

Rätvinklig triangel: en vinkel rät

Pythagoras sats: $a^2 + b^2 = c^2$

Bisektrissatsen: $x/y = a/b$

Triangelns märkliga punkter
\begin{enumerate}
    \item Mittpunktsnormaler (omskrivna cirkelns medelpunkt)
    \item Bisektriser (inskrivna cirkelns medelpunkt)
    \item Höjder
    \item Median (förhållandet 2:1)
\end{enumerate}

\subsection{Fyrhörning}
Trapets: en fyrhörning som har två parallella sidor, medan de två övriga sidorna inte är parallella
\[
A = \frac{a+b}{2}h
\]

Parallellogram: en fyrhörning vars sidor är parvis parallella. De motsåtende vinklarna i en parallellogram är sinsemellan lika stora. De motstående sidorna är lika långa och diagonalerna halverar varandra.
\[
A = ah
\]

Romb: en parallellogram vars alla sidor är lika långa

Rektangel: en parallellogram vars alla vinklar är räta

Kvadrat: all sidor lika stora och vinklarna räta

Rombens diagonaler halverar varandra och ör vinkelröta

\subsection{Regelbunden månghörning}
Alla sidor lika långa och alla vinklar lika stora

\section{Absolutbelopp}
\subsection{Definition}
\[
|a| = \begin{cases}a & a \geq 0 \\ -a & a < 0 \end{cases}
\]

\begin{example}
$-x|10-5x| + 5x +1 $

$\begin{cases}
-x(10-5x) + 5x +1, 10-5x \geq 0
-x(-10+5x) + 5x +1, 10 - 5x < 0
\end{cases}$
\end{example}

\subsection{Ekvationer}
$|a| = b \iff a = b \lor a = -b$

$|a| = |b| \iff a=b \lor a = -b$

\begin{example}
$|2x-4|=|5-x|-x$

$x < 2 -(2x-4) = 5 -x -x$

$2 \leq x \leq 5 2-x = 5-x-x$

$x > 5 2x - 4 = -(5-x)-x$
\end{example}

\subsection{Olikheter}
$|a| < b$, $-b  < a < b$

$|a| > b \iff a > b$ or $a < -b$

$|a| < |b| \iff a^2 < b^2$

\begin{example}
$|-3x+6|>|x|+2$

1. $x<0 -3x +6 > -x + 2$ $to$ $x < 0$

2. $x \leq x \leq 2$, $-3x + 6 > x +2$, $x < 1$, $0 \leq x > 1$

3. $x>2$ and $3x -6 > x +2$, $x > 4$
\end{example}

\subsectio{En linjes lutning}
Bestäms som den riktade vinkeln mellan linjen och x-axeln. Denna riktade vinkel $\alpha$ kallas linjens riktningsvinkel $-90^o < \alpha \leq 90^o$

Riktningskoefficienten
\[
K = \frac{\Delta y}{\Delta x} = \frac{y_2 - y_1}{x_2 - x_1}
\]
då $(x_1, y_1)=P_1$ ja $(x_2, y_2) = P_2$

\begin{enumerate}
    \item $k>0$ stigande linje
    \item $k < 0$ fallande linje
    \item $k=0$ vågrät linje
    \item $k$ odefinierad, lodrät linje
\end{enumerate}

$\tan \alpha = k$

\subsection{Cirkel och linje}

Tangent genom punkt på cirkeln:

Tangent:
\[
y-(-1) = k(x - (-3)) \iff kx - y + 3k -1 = 0
\]

\[
\frac{|k-2+3k-1|}{\sqrt{k^2 + (-1)^2}} = 5 \iff k=-4/3

Tangent:
\[
y-1 = k(x - (-3)) \iff kx - y + 3k + 1 = 0
\]
Avstånd
\[
\frac{|k\cdot0 - 1\cdot 0 + 3k + 1|}{\sqrt{k^2 + (-1)^2}} = \sqrt{5}
\]
$k = 1/2$ eller $k = -2$

Linjer: $x - 2y -5 = 0$ och $2x + y + 5= 0$

\section{Parabeln}
\subsection{Definition}
Parabel: mängden av de punkter i planet som har samma avstånd till en fast linje och till en fast punkt utanför linjen 

\subsection{Parabel som pönnar sig uppåt eller nedåt}
Allmän form: $y = ax^2 + bx + c$

Parabelns topp: $y - y_0 = a(x-x_0)^2$

Parabelns nollställen: $Y = a(x-x_1)(x - x_2)$

Parabelns topp: $x_0 = -b/2a$

\subsection{Parabel som öppnar sig åt höger eller vänster}
Allmän form: $x = ay^2 + by + c$

Parabelns topp: $x - x_0 = a(y-y_0)^2$

Parabelns nollställen: $x = a(y-y_1)(y-y_2)$

Parabelns topp: $y_0 = -b/2a$
