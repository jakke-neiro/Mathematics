\section{Taustatietojen kertausta ja täydennystä}
\subsection{Käytännön joukko-oppia}
Olkoon $X$ mikä tahansa joukko. Tällöin $X$:n potenssijoukko on
\[
P(X) = \{A: A \subset X\}
\]
ja $X$:n joukkoperhe (tai perhe/kokoelma X:n osajoukkoja) on mikä tahansa $P(X):n$ osajoukko
\[
F \subset P(X)
\]
Perheen $F$ yhdiste on 
\[
\bigcup_{A \in F}A = \{ x \in X: x \in A \textrm{jollakin} A \in F\}
\]
ja leikkaus on
\[
\bigcap_{A \in F}A = \{x \in X: x \in A \textrm{kaikilla} A \in F \}
\]
Olkoon $A$ jokin (indeksi)joukko ja oletetaan, että jokaista $\alpha \in A$ vastaa yksikäsitteinen osajoukko $V_{\alpha} \subset X$. Toisin sanoen, $\alpha \to V_{\alpha}$ on kuvaus $A \to P(X)$. Tällöin kokoelma
\[
F = \{V_{\alpha}: \alpha \in A\}
\]
on $X$:n indeksöity joukkoperhe.

Indeksöidyn perheen yhdiste on
\[
\bigcup_{\alpha \in A}V_{\alpha} = \{x \in X: x \in V_{\alpha} \textrm{jollakin} \alpha \in A \}
\]
ja vastaavasti leikkaus on
\[
\bigcap_{\alpha \in A}V_{\alpha} = \{x \in X: x \in V_{\alpha}\textrm{kaikilla}\alpha \in A\}
\]
Merkitsemme myös 
\[
\bigcup_{\alpha}V_{\alpha} ja \bigcap_{\alpha}V_{\alpha}
\]
jos $A$ käy selville asiayhteydestä.

\subsubsection{Esimerkki}
Olkoon $F \subset P(X)$. Voimme tulkita $F$:n indeksöidyksi perheeksi käyttämällä $F$:ää itseään indeksijoukkona. Toisin sanoen, jos $\alpha \in F$ (jolloin $\alpha$ on $X$:n osajoukko) merkitään $V_{\alpha} = \alpha$. Tällöin $F = \{V_{\alpha}: \alpha \in F\}$.
\[
X = \bigcup_{x \in X}\{x \}, \{x\}=\textrm{yksiö}
\]
Usein indeksijoukkona on $\N = \{1,2,3,\dots\}$, jolloin merkitään
\[
\bigcup_{n \in \N}V_n, \bigcup_n^{\infty}V_n, \bigcup_n V_n
\]
ja vastaavasti
\[
\bigcap_{n \in \N}V_n, \bigcap_n^{\infty}V_n, \bigcap_n V_n
\]
Merkinnät $(V_n)$, $(V_n)_{n=1}^{\infty}$, $(V_n)_{n \in \N}$ ja $V_1, V_2, \dots$ tarkoittavat (joukkojen) jonoja. Joukkojen $A, B \subset X$ erotus on
\[
A \setminus B = \{x \in X: x \in A \textrm{ja} x \notin B\}
\]
Joukon $B \subset X$ komplementti ($X$:n suhteen) on
\[
B^{C} = X \setminus B
\]

Olkoon $\{V_{\alpha}: \alpha \in A\}$ jokin $X$:n joukkoperhe. Tällöin pätee ns. de Morganin lait. 
\[
\bigg( \bigcup_{\alpha}V_{\alpha} \bigg)^{C} = \bigcap_{\alpha} V_{\alpha}^{C}
\]
ja
\[
\bigg(\bigcap_{\alpha} V_{\alpha} \bigg)^{C} = \bigcup_{\alpha} V_{\alpha}^{C}
\]
Olkoon $B \subset X$. Tällöin pätee ns. distributiiviset lait yhdisteelle ja leikkaukselle:
\[
B \cap \bigg(\bigcup_{\alpha}V_{\alpha}) = \bigcap_{\alpha}(B \cap V_{\alpha})
\]
ja
\[
B \cup \bigg( \bigcap_{\alpha}V_{\alpha} \bigg) = \bigcap_{\alpha}(B \cup V_{\alpha}) 
\]
Todistus
\[
x \in \bigg(\bigcup_{\alpha}V_{\alpha} \bigg)^{C} \iff x \notin \bigcup_{\alpha}V_{\alpha} \iff \forall \alpha: x \notin V_{\alpha} \iff \forall \alpha : x \in V_{\alpha}^{C} \iff x \in \bigcap_{\alpha}V_{\alpha}^{C}
\]
\[
x \in B \cap (\bigcup_{\alpha}V_{\alpha}) \iff x \in B \textrm{ja} x \in \bigcup_{\alpha}V_{\alpha} \iff x \in B \textrm{ja} x \in V_{\alpha} \textrm{jollakin} \alpha \in A \iff x \in B \cap V_{\alpha} \textrm{jollakin} \alpha \in A \iff x \in \bigcup_{\alpha}(B \cap V_{\alpha})
\]

Joukkoperheen yhdisteen/leikkauksen kuvat ja alkukuvat. Olkoot $X$ ja $Y$ epätyhjiä joukkoja ja $f: X \to Y$ kuvaus. Joukon $A \subset X$ kuva (kuvauksessa $f$) on
\[
f(A) = \{f(x): x \in A \} (\subset Y)
\]
Merkitään myös lyhyemmin $fA$.
Joukon $B \subset Y$ alkukuva (kuvauksessa $f$) on
\[
f^{-1}(B) = \{x \in X: f(x) \in B \}
\]
Merkitään myös $f^{-1}B$. Käytämme myös merkintää
\[
f^{-1}(y) = f^{-1}(\{y\})
\]
kun $y \in Y$. $f$:llä ei tarvitse olla käänteiskuvausta.

Olkoon $f: X \to Y$, $\{V_{\alpha}: \alpha \in A \}$ $X$:n joukkoperhe ja $\{W_{\beta}: \beta \in B\}$ Y:n joukkoperhe. Silloin
\[
f(\bigcup_{\alpha}V_{\alpha}) = \bigcup_{\alpha}fV_{\alpha}
\]
\[
f^{-1}(\bigcup_{\beta}W_{\beta}) = \bigcup_{\beta}f^{-1}W_{\beta}
\]
\[
f^{-1}(\bigcap_{\beta}W_{\beta}) = \bigcap_{\beta}f^{-1}W_{\beta}
\]
Todistus
\[
y \in f(\bigcup_{\alpha}V_{\alpha}) \iff y = f(X) \textrm{ja} x \in \bigcup_{\alpha}V_{\alpha} \ff y=f(x) \textrm{ja} x \in V_{\alpha} \textrm{jollakin}\alpha \in A \iff
y \in f V_{\alpha} \textrm{jollakin} \alpha \in A \iff y \in \bigcup_{\alpha}f V_{\alpha}
\]

Aina pätee
\[
f(\bigcap_{\alpha}V_{\alpha}) \subset \bigcap_{\alpha}f V_{\alpha}
\]

\subsubsection{Numeroituvat ja ylinumeroituvat joukot}
Numeroituvuus on erittäin tärkeä mittateoriassa! 

Joukko $A$ on numeroituva, jos $A = \emptyset$ tai $\exists$ injektio $f: A \to \N (\iff \exists$ surjektio $g: \N \to \A)$. $A$ on ylinumeroituva, jos $A$ ei ole numeroituva. 

$A$ on numeroituva $\iff$ $A$ äärellinen (mukaanluettuna $\emptyset$) tai numeroituvasti ääretön (jolloin $\exists$ bijektio $f: A \to \N$).

$A \neq \emptyset$ $\iff$ $A = \{x_n: n \in \N\}$ (toisto sallittu, joten $A$ voi olla äärellinen).

$A$ numeroituva, $B \subset A \to B$ numeroituva.

Jos joukot $A_n$ ovat numerootuvia $\forall n \in \N$, niin
\[
\bigcup_{n \in \N}A_n
\]
on numeroituva. Eli numeroituva yhdiste numeroituvista joukoista on numeroituva. 

Voi olettaa $A_n \neq \emptyset$ $\forall n \in \N$. $A_n$ numeroituva $\to$ $A_n = \{x_m(n): m \in \N \}.$ Määritellään kuvaus 
\[
g: \N \times \N \to \bigcup_n A_n, g(n, m) = x_m(n)
\]
Silloin $g$ on surjektio $\N \times \N \ti \bigcup_n A_n$. Riittää löytää surjektio $h: \N \to \N \times \N$, koska silloin
\[
g \circ h: \N \to \bigcup_{n \in \N}A_n
\]
on surjektio ja siten $\bigcup_n A_n$ on numeroituva. Esimerkki surjektiosta $h: \N \to \N \times \N$.

Rationaalilukujen joukko
\[
\Q = \{\frac{m}{n} | n, m \in \Z, n \neq 0 \}
\]
on numeroituva. Syy: Joukko
\[
A_k = \{ \frac{m}{n} | n,m \in \Z, n \neq 0, |m| \leq k, |n| \leq k \}
\]
on äärellinen (ja siten numeroituva) $forall k \in \N$. 

(Ylinumeroituva joukko). Väli $[0, 1]$ (ja siten myös $\R$) on ylinumeroituva. Idea: $x \in [0, 1] \to$ $x$:llä on desimaalikehitelmä
\[
x = 0, a_1 a_2 a_3 \dots
\]
missä $a_j \in \{0,1,2,\dots,9\}$.

Vastaoletus: $[0,1]$ numeroituva, jolloin $[0,1] = \{x_n: n \in \N \}$. Pisteillä $x_n$ on desimaalikehitelmät
\[
x_1 = 0, a_1^{(1)}a_2^{(2)}a_3^{(1)}\dots
\]
\[
x_2 = 0, a_1^{(2)}a_2^{(2)}a_3^{(2)}\dots
\]
\[
x_3 = 0, a_1^{(3)}a_2^{(3)}a_3^{(3)}\dots
\]
\[
x_n = 0, a_1^{(n)}a_2^{(n)}a_3^{(n)}\dotsa_n^{(n)}\dots
\]
Lävistäjällä on lukujono $a_1^{(1)}, a_2^{(2)}, a_3^{(3)}, \dots, a_n^{(n)}, \dots$ missä $a_n^{(n)} = x_n$:n $n$:s desimaali. Määritellään luku $x \in [0, 1]$ asettamalla $x = 0,b_1 b_2 b_3 \dots$, missä 
\[
b_n = \begin{cases}
a_n^{(n)} + 2, \textrm{jos} a_n^{(n)} \in \{0,1,2,\dots,7\} \\
a_n^{(n)}-2, \textrm{jos} a_n^{(n)} \in \{8,9\}
\end{cases}
\]
Luvun $x$ $n$:s desimaali toteuttaa $|b_n - a_n^n| = 2 \forall n \in \N$, joten $x \neq x_n \forall n \in \N$. Tämä on ristiriita, sillä $[0,1] = \{ x_n: n \in \N \}$. Siis $[0,1]$ on ylinumeroituva.

Summaaminen. Olkoon $A \neq \emptyset$ mielivaltainen indeksijoukko ja $a_{\alpha} \geq 0 \forall \alpha \in A$. 
\[
\sum_{\alpha \in A}a_{\alpha} = \sup \{ \sum_{\alpha \in A_0} a_{\alpha} | A_0 \subset A \textrm{äärellinen} \}
\]

\subsection{Euklidinen avaruus}
\[
\R^n = \R \times \dots \times \R
\]
karteesinen tulo. Alkioita kutsutaan pisteiksi tai vektoreiksi. 
\[
x \in \R^n \iff x = (x_1, \dots, x_n), x_j \in \R, j=1,\dots, n
\]
Algebrallinen rakenne

Pisteiden $x, y \in \R^n$ summa on
\[
x + y = (x_1 + y_1, \dots, x_n + y_n)\in \R^n
\]
Reaaliluvun $\lambda \in \R$ ja pisteen $x \in \R^n$ tulo on
\[
\lambda x = (\lambda x_1, \dots, \lambda x_n) \in \R^n
\]
Nollavektori
\[
0 = (0, \dots, 0)
\]
Pisteen $x \in \R^n$ vastavektori (vastinpiste)
\[
-x = (-1)x = (-x_1, \dots, -x_n)
\]
Pisteiden $x, y \in R^n$ erotus on
\[
x-y = x + (-y)
\]
$\R^{n}$:ssä summa ja reaaliluvulla kertominen toteuttavat vektoriavaruuden ehdot (lin. alg. I), esim.
\[
x + y = y + x, x + 0 = 0 + x = x
\]
\[
\lambda(x + y) = \lambda x + \lambda y, (\lambda + \mu)x = \lambda x + \mu x
\]
\[
\forall x, y \in \R^{n}, \lambda, \mu \in \R
\]
Pisteiden $x, y \in \R^n$ sisätulo on
\[
x \cdot y = \sum_{i = 1}^n x_i y_i \in \R
\]
Merkitään 
\[
|x| = \sqrt{x \cdot x} = \bigg(\sum_{i =1}^n x_i x_i \bigg)^{1/2}
\]
$x$:n normi. 

Euklidinen etäisyys $\R^{n}$:ssä. Pisteiden $x, y \in \R^n$ etäisyys on
\[
|x - y| = \bigg( \sum_{i=1}^n (x_i - y_i)^2 \bigg)^{1/2}
\]
Usein merkitään $d(x, y) = |x - y|$. Tällöin $d$ on metriikka $\R^{n}$:ssä, ts. kuvaus $d: \R^n \times \R^n \to \R$ toteuttaa metriikan ehdot:
\[
d(x, y) \geq 0 \forall x, y \in \R^n
\]
\[
d(x, y) = 0 \iff x = y
\]
\[
d(x, y) = d(x, y) \forall x, y \in \R^n
\]
\[
d(x, y) \leq d(x, z) + d(z, y) \forall x, y, z \in \R^n
\]

Avoimet ja suljetut joukot $\R^n$:ssä. Euklidinen metriikka $d$ määrää $\R^n$:ään avoimet ja suljetut joukot (ja siten $\R^n$:n topologian) seuraavasti:

Olkoon $x \in \R^n$ ja $r > 0$. Joukko
\[
B(x, r) = \{y \in \R^n: | y - x | < r \}
\]
on avoin (x-keskeinen, r-säteinen) kuula (ball) ja
\[
S(x, r) = \{y \in \R^n: |y - x| = r \}
\]
on (x-keskinen, r-säteinen) pallo (tai pallokuori) (engl. "sphere"). Vastaavasti
\[
\overline{B}(x, r) = \{y \in \R^n: |y - x| \leq r \}
\]
on suljettu (x-keskinen, r-säteinen) kuula.

Joukko $V \subset \R^n$ on avoin, jos $\forall x \in V \exists r=r(x) > 0$ sellainen että $B(x, r) \subset V$. Joukko $V \subset \R^n$ on suljettu, jos $\R^n \setminus V$ on avoin.

$B(x, r)$ on avoin $\forall x \in \R^n, r> 0$. 

Suljettu kuula $\overline{B}(x, r)$ on suljettu joukko. 

$\R^n$ ja $\emptyset$ ovat sekä avoimia että suljettuja.

Puoliavoin väli, esim. $[0,1)$, ei ole avoin eikä suljettu.

Joukon $A \subset \R^n$ sulkeuma on
\[
\overline{A} = \{x \in \R^n: x \in A \textrm{tai} x \textrm{on A:n kasautumispiste} \}
\]
Piste $x \in \R^n$ on joukon $A \subset \R^n$ kasautumispiste, jos $\forall r > 0 B(x, r) \cap (A \setminus \{x\}) \neq \emptyset$. $\R^{n}$:ssä pätee $\overline{B}(x, r) = B(x, r)$.

Joskus kuulaa $B(x, r)$ kutsutaan myös palloksi, jolloin $S(x, r)$:ää on kutsuttava pallokuoreksi.

Jos $(X, d)$ on metrinen avaruus, ts. $d: X \times X \to \R$ toteuttaa metriikan ehdot, niin voidaan määritellä $X$:n avoimet ja suljetut joukot (metr. $d$ suhteen) kuten edellä korvaamalla $|y - x|$ metriikalla $d(x, y)$. 

Olkoon $A$ mikä tahansa indeksijoukko. Silloin
\[
V_{\alpha} \subset \R^n avoin \forall a \in A \to \bigcup_{\alpha \in A}V_{\alpha} avoin
\]
\[
V_{\alpha} \subset \R^n \forall \alpha \in A \to \bigcap_{\alpha \in A}V_{\alpha} suljettu
\]
\[
V_1, \dots, V_k \subset \R^n avoin \to \bigcap_{j = 1}^k V_j avoin
\]
\[
V_1, \dots, V_k \subset \R^n suljettu \to \bigcup_{j=1}^k V_j suljettu
\]
Todistus
\[
x \in \bigcup_{\alpha \in A}V_{\alpha} \to \exists \alpha_0 \in A \textrm{sellainen että} x \in V_{\alpha_0}
\]
\[
V_{\alpha_0}\textrm{avoin}\to \exists \textrm{avoin kuula}B(x, r) \subset V_{\alpha_0} \subset \bigcup_{\alpha \in A}V_{\alpha}
\]
\[
V_{\alpha}\textrm{suljettu}\forall \alpha \to V_{\alpha}^{C}\textrm{avoin}\forall \alpha
\]
\[
\to \bigcup_{\alpha}V_{\alpha}^{C} = (\bigcap_{\alpha} V_{\alpha})^{C}\textrm{avoin}
\]
\[
\to \bigcap_{\alpha}V_{\alpha}\textrm{suljettu}
\]

\section{Lebesguen mitta $\R^n$:ssä}
\subsection{Johdanto}
Geometrinen lähtökohta. Jos $I = [a, b] \subset \R$ on rajoitettu väli, niin sen pituus on
\[
l(I) = b - a
\]
Joukko $I \subset \R^n$ on n-väli, jos se on muotoa 
\[
I = I_1 \times \dots \times I_n
\]
missä kukin $I_j \subset \R$ on väli (joko avoin, suljettu tai puoliavoin).

$I$ on avoin (vastaavasti suljettu) n-väli, jos jokainen $I_i$ on avoin (vastaavasti suljettu). Olkoot
