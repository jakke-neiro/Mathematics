\section{Laskutoimitus}
Algebra tuktii laskutoimituksia, niiden ominaisuuksia ja niistä muodostuvia rakenteita

Määritelmä: Joukon $S$ laskutoimitus $*$ on sääntö, josa liittää jokaiseen joukon $S$ alkiopariin $(x, y)$ jonkin yksikäsitteisen kolmannen alkion joukosta $S$. Tätä alkiota kutsutaan laskutoimituksen tulokseksi $x * y$

Kuvaus: $S \times S \to S, (x, y) \to x * y$ Jokaista paria $(x, y)$ vastaa jokin alkio joukosta $S$

Tuloksen on oltava yyksikäsitteinen, Pari $(S, *)$

\begin{example}
Yhteenlasku on joukon $\N$ laskutoimitus
\end{example}

\subsection{Liitännäisyys ja vaihdannaisuus}
Määritelmä: Liitännäinen: jos $x*(y*z) = (x*y)*z \forall x, y, z \in S$

Vaihdannainen: jos $x*y = y*x \forall x, y \in S$

Liitännäisyys: assosiatiivisuus, Vaihdannaisuus: kommutatiivisuus

Kummutoivat alkiot: $a, b \in S$, $a * b = b * a$

\subsection{Neutraali- ja käänteisalkiot}
Määritelmä: Joukon $S$ alkiota $e$ kutsutaan neutraalialkioksi, jos $e * x = x$ ja $x * e = x \forall x \in S$

Lause: Jos laskutoimituksella $*$ on neutraalialkio, se on yksikäsitteinen. 

Määritelmä: Oletetaan, että laskutoimituksella $*$ on enutraalialkio $e$. Olkoon $x \in S$. Alkiota $x'$ kutsutaan alkion $x$ käänteisalkioksi, jos $x*x' = e$ ja $x'*x = e$

Alkiota, jolla on käänteisalkio, nimitetään kääntyväksi.

Lause: Oletetaan, että laskutoimitus $*$ on liitännäinen, ja että sillä on neutraalialkio $e$. Jos alkiolla $x \in S$ on käänteisalkio, se on yksikäsitteinen. 
 
\subsection{Laskutoimitustaulu}
Äärellinen joukko

Symmetrinen lävistäjän suhteen $\to$ vaihdannainen

Algebra: symbolinen laskenta

\section{Ryhmä}
Joukko yhdellä tai useammalla laskutoimituksella $\to$ algebrallinen rakenne 

Aksiomaattinen määrittely: luetellaan joukko ehtoja

Määritelmä: Joukko $G$ laskutoimituksella $*$ varustettuna on ryhmä, jos seuraavat ehdot ovat voimassa
\begin{enumerate}
    \item Laskutoimitus $*$ on määritelty joukossa $G$, eli kaikilla alkioilla $x, y \in G$ pätee $x*y \in G$
    \item Laskutoimitus $*$ on liitännäinen
    \item Joukossa $G$ on neutraalilkio laskutoimitukselle $*$
    \item Jokaisella $G$:n alkiolla on käänteisalkio laskutoimituksen $*$ suhteen
\end{enumerate}

"G on ryhmä" $\to (G, *)$ neutraalilkio, käänteisalkio yksikäsitteisiä

Abelin ryhmä = vaihdannainen ryhmä

Ryhmän $G$ kertaluku $|G| = #G = \cird(G)$ on alkioiden lukumäärä

Lause: Oletetaan, että $(G, *)$ on ryhmä ja että $a, b \in G$. Tälllöin yhtälöllä $a*x = b$ on yksikäsitteinen ratkaisu, samoin yhtälöllä $x * a = b$

\subsection{Merkintätavoista}
Ryhmäteoriassa on tapana merkitä ryhmälaskutoimitusta tähden sijaan kertomerkikkä ($\cdot$). Tulos = tulo, G on ryhmä = $(G, \cdot)$

Neutraalialkio $e_{G}$, jossa $G$ on ryhmä. Käänteisalkio $a^{-1}$

\subsection{Potenssi}
Ryhmän $G$ potenssi $n \geq 1$, kokonaisluku

$x^n = x \cdot x \cdot \dots \cdot x$, n kpl, $x^0 = e$, $x^{-n} = (x^n)^{-1}$

Lause: Olkoon $G$ ryhmä ja $x$ sen alkio. Seuraavat kaavat pätevät kaikilla kokonaisluvuila $m$ ja $n$
\[
(x^n)^{-1} = (x^{-1})^n
\]
\[
X^m \cdot x^n = x^{m+n} 
\]
\[
(x^n)^m = x^{mn}
\]

\subsection{Monoidit ja puoliryhmät}
Määritelmä: Pari $(M, *)$ on monoidi, jos euraavat ehdot ovat voimassa:
\begin{enumerate}
    \item Joukossa $M$ on määritelty laskutoimitus $*$
    \item Laskutoimitus $*$ on liitännäinen
    \item Joukossa $M$ on neutraalialkio laskutoimitukselle $*$
\end{enumerate}

Puoliryhmä: toteuttaa vain ehdot $M1$ ja $M2$.

Ryhmät $\subset$ Monoidit $\subset$ Puoliryhmät $\subset$ Algebralliset rakenteet

\subsection{Ryhmien laskutoimitustaulut}
Lemma: Äärellisen ryhmän laskutoimitustaulussa jokainen alkioista esiintyy täsmälleen kerran jokaisella rivillä ja jokaisessa sarakkeessa. 

\subsection{Aliryhmä}
Määritelmä: Oletetaan, että $(G, \cdot)$ on ryhmä ja $H \subset G$. Pari $(H, \cdot)$ on ryhmän $(G, \cdot)$ aliryhmä, jos seuraavat ehdot toteutuvat
\begin{enumerate}
    \item Joukko $H$ on vakaa laskutoimituksen suhteen, eli jos $g, h \in H$, niin $gh \in H$
    \item Ryhmän $(G, \cdot)$ neutraalialio on joukossa $H$.
    \item Joukko $H$ sisältää kaikkien alkioidensa käänteisalkiot laskutoimituksen $\cdot$ suhteen, eli $g^{-1} \in H \forall g \in H$
\end{enumerate}

$(H, \cdot) \leq (G, \cdot)$ tai $h \leq G$, $H \neq G \to H < G$ aito aliryhmä

Lause: Oletetaan, että $H_1 \leq G$ ja $H_2 \leq G$. Tällöin $H_1 \cap H_2 \leq G$

Lause: Oletetaan, että $(G, \cdot)$ ja $(H, \cdot)$ ovat ryhmiä. Jos $H$ on joukon $G$ osajoukko, se on ryhmän $G$ aliryhmä-

Lause: (Aliryhmäkriteeri). Oletetaan, että $H$ on ryhmän $G$ epätyhjä osajoukko. Tällöin $H$ on ryhmän $G$ aliryhmä, jos ja vain jos $ab^{-1} \in H \forall a, b \in H$.

\section{Ryhmien isomorfia}
Ryhmät rakenteeltaan samanlaisia

\subsection{Kertotaulujen vertailua}
Äärellisellä ryhmällä samanlainen laskutoimitustaulu $\to$ rakenteeltaan samanlaiset

$G = \{e, a, b\}$ ja $H = \{x, y, z\}$

\subsection{Kolmi- ja nelialkioiset ryhmät}
Lause: Kolmialkioisia ryhmiä on vain yksi, jos isomorfiset ryhmät lasketaan samoiksi

Lause: $N$ erialkioisia ryhmiä on täsmälleen kaksi kappaletta, jos isomorfiset ryhmät lasketaan samoiksi.

Kaikki ryhmän alkiot voidaan esittää alkion $a$ potensseina $\{a, a^2, a^3, a^4\}$

\subsection{Isomorfismi}
Bijektio: $f: G \to H$, $a, b \in G \to ab \to f(a)f(b)$

Määritelmä: Olkoot $(G, *)$ ja $(H, \circ)$ ryhmiä ja $f: G \to H$ jokin kuvaus. Kuvaus $f$ on ryhmäisomorfismi tai lyhyemmin isomorfismi, jos seuraavat ehdot pätevät
\begin{enumerate}
    \item Kuvaus $f$ on bijektio
    \item Kaikilla ryhmän $G$ alkioilla $x, y$ pätee $f(x*y) = f(x)\circ f(y)$
\end{enumerate}

$(G, *) \equiv (H, \circ)$

\subsection{Isomorfismin ominaisuuksia}
Lause: Jos $f:(G, *) \to (H, \circ)$ on ryhmäisomorfismi, seuraavat väitteet pätevät $f(e_G) = e_H$, $f(g^{-1}) = f(g)^{-1}$, $g \in G$

Lemma: Olkoon $f: G \to H$ ryhmäisomorfismi ja olkoon $g \in G$. Tällöin kaikilla $k \in \Z$ pätee $f(g^k) = f(g)^k$

Lause: Olkoon $f: G \to H$ isomorfismi. Tällöin käänteiskuvaus $f^{-1}$ on myös isomorfismi.

Lause: Olkoot $f: G \to H$ ja $g: H \to K$ ryhmäisomorfismeja. Tällöin yhdistetty kuvaus $g \circ f: G \to K$ on myös ryhmäisomorfismi.

Korollaari: Olkoot $G, H$ ja $K$ ryhmiä. Seuraavat säännöt pätevät
$G \equiv G$, $G \equiv H \to H \equiv G$, $G \equiv H \land H \equiv K \to G \equiv K$

\section{Virittäminen}

\subsection{Yhden alkion virittämä aliryhmä}
$A_3 = \{(123), (123)^2, (123)^3 \} = \{(123), (132), (1) \}$

Aliryhmä $A_3$ on pienin aliryhmä, joka sisältää alkion 123

Määritelmä: Olkoon $G$ ryhmä ja $g$ sen alkio. Pienintä aliryhmää, joka sisältää alkion $g$, kutsutaan alkion $g$ virittämäksi aliryhmäksi ja merkitään symbolilla $\langle g \rangle$. Alkiota $g$ kutsutaan aliryhmän $\langle g \rangle$ virittäjäksi.

\begin{example}
Ryhmän $(Q^^', \cdot)$ alkion $-1$ virittämä aliryhmä $\langle -1 \rangle = \{(-1)^n|n\in \Z \}$
\end{example}

\subsection{Kertaluku}
Alkion kertaluvuksi kutsutaan sen virittämän aliryhmän alkioiden lukumäärää

Määritelmä: Olkoon $G$ ryhmä. Alkion $g \in G$ kertaluku $O(g)$ on sen virittämän aliryhmän $\langle r \rangle$ kertaluku

Lemma: Olkoon $G$ ryhmä ja $g$ jokin sen alkio. Oletetaan, että positiiviselle kokonaisluvulle $n$ pätee $g^n = e$. Tällöin $\langle g \rangle = \{e, g, g^2, \dots, g^{n-1}\}$

Lause: Olkoon $G$ ryhmä ja $g$ jokin sen alkio. Alkion $g$ kertaluku on pienin positiivinen kokonaisluku $n$, jolla pätee $g^n = e$. Jos tällaista lukua ei löydy, kertaluku on ääretön.

Lause: Jos alkion $g$ kertaluku on ääretön, sen virittämä aliryhnä on $\langle g \rangle = \{g^k | k \in \Z \}, g^k \neq g^n, k \neq m$

\subsection{Useamman alkion virittämä aliryhmä}
Määritelmä: Olkoon $G$ ryhmä ja olkoon $S \subset G$. Joukon $S$ virittämä aliryhmä $\langle S \rangle$ on pienin ryhmän $G$ aliryhmä, joka sisältää joukon $S$.

\begin{enumerate}
    \item Merkitään arvattua joukkoa vaikkapa symbolilla $H$ ja osoitetaan, että $S \subset H$
    \item Osoitetaan, että $H$ on aliryhmä
    \item Oletetaam että $K$ on jokin joukon $S$ sisältävä aliryhmä, ja osoitetaan, että $H \subset K$.
\end{enumerate}

\begin{example}
$S_4$, $S = \{(13), (24) \}$

$(13)^2 = 1$, $(24)^2 = (1)$, $(13)\cdot(24) = (13)(24)$, $(24) \cdot (13) = (24)(13)$

$(13)^{-1} = (13)$, $(24)^{-1} = (24)$
\end{example}

\section{Lukuteoria}
Lukuteoria $\to$ kertaluvut $\to$ virittäminen $\to$ sykliset ryhmät

\subsection{Jaollisuus}
Määritelmä: Kokonaisluku $n$ on jaolinen kokonaisluvulla $m$, jos jollain kokonaisluvulla $a$ pätee $n=am$. Tällöin merkitään $m \mid n$

$m$ on luvun $n$ tekijä tai $n$ on luvun $m$ monikerta, $m \not \mid n$ 

Lause: Jakoyhtälö. Olkoot $a$ ja $b$ kokonaislukuja. Oletetaan, että $b \neq 0$. Tällöin on olemassa yksikäsitteiset $q, r \in \Z$, joille pätee
\[
a = qb +r, 0 \leq r < |b|
\]
$r$ on luvun $a$ jakojäännös luvulla $b$ jaettaessa.

\subsection{Suurin yhteinen tekijä}
Määritelmä: Olkoot $a$ ja $b$ kokonaislukuja, joista vähintään toinen on nollasta poikkeasva. Suurinta lukua, joka jakaa sekä $a$:n että $b$:n, kutsutaan niiden suurimmaksi yhteiseksi tekijäksi ja merkitään $syt(a, b)$. 

Lause: (Bezout'n lause). Olkoot $a$ ja $b$ kokonaislukuja, joista toinen on nollasta poikkeava. Tällöin on olemassa kokonaisluvut $x$ ja $y$, joille pätee $syt(a, b) = xa + yb$. 

Korollaari: Oletetaan, että $a$ ja $b$ ja $c$ ovat kokonaislukuja, joista joko $a$ tai $b$ on nollasta poikkeava. Yhtälöillä $ax + by = c$ on kokonaislukuratkaisu, jos ja vain jos $syt(a, b)|c$

Lemma: Euklideen lemma. Oletetaan, että $a, b, c \in \Z \setminus \{0\}$ ja että $a$ ja $b$ ovat keskenään jaottomat. Jos $a\mid bc$, niin $a \mid c$. 

Määritelmä: Olkoot $a$ ja $b$ nollasta poikkeavia kokonaislukuja. Pienintä positiivista lukua, jonka sekä $a$ että $b$ jakavat, kutsutaan niiden pienimmäksi yhteiseksi jaettavaksi (tai monikerraksi) ja merkitään $pyj(a, b)$. 

Lause: Nollasta poikkeaville kokonaisluvuille $a$ ja $b$ pätee
\[
syt(a, b)pyj(a, b) = ab
\]

\subsection{Alkuluvut}
Määritelmä: Kokonaisluku $p > 1$ on alkuluku, jos sen ainoat positiiviset tekijät ovat $1$ ja $p$.

Lause: Olkoot $a, b \in \Z$, ja olkoon $p$ alkuluku. Jos $p \mid ab$, niin $p \mid a$ tai $p \mid b$. 

Korollaari: Olkoot $a_1, a_2, \dots, a_n \in \Z$, ja olkoon $p$ alkuluku. Jos $p$ jakaa tulon $a_1a_2 \dots a_n$ , niin $p$ jakaa jonkin luvuista $a_i$. 

Lause: Arimetiikan peruslause. Jokainen positiivinen kokonaisluku voidaan esittää alkulujune tulona. Esitys on tekijöiden järjestystä lukuunottamatta yksikäsitteinen. 

Lause: Alkulukuja on äärettömän monta.

\subsection{Kongruenssi}
Lemma: Luvuilla  $a$ ja $b$ on sama jakojäännös luvulla $n$ jaettaessa, jos ja vain jos $n \mid (a-b)$. 

Määritelmä: Oletetaan, että $a, b \in \Z$ ja $n$ on positiivinen kokonaisluku. Jos $a-b$ on jaollinen luvulla $n$, sanotaan, että $a$ ja $b$ ovat kongruentit modulo $m$ ja merkitään $a \equiv b$ (mod $n$)

Lemma: Olkoot $a,b, c \in \Z$, ja olkoon $n$ positiivinen kokonaisluku. Tällöin seuraavat väitteet pätevät
\begin{enumerate}
    \item $a \equiv b$ (mod $n$)
    \item Jos $a \equiv b$ (mod $n$), niin $b \equiv a$ (mod $n$)
    \item Jos $a \equiv b$ (mod $n$) ja $b \equiv c$ (mod $n$), niin $a \equiv c$ (mod $n$)
\end{enumerate}

Määritelmä: Olkoon $n$ positiivinen kokonaisluku. Kokonaisluvun $a$ jäännösluokka modulo $n$ on joukko
\[
[a]_n = \{b \in \Z|b \equiv a (mod n) \}
\]
Lukua $a$ kutsutaan jäännösluokan $[a]_n$ edustajaksi

\subsection{Kongruenssilla laskeminen}
Lause: Olkoot $a, b, c, d \in \Z$ ja olkoon $n$ positiivinen kokonaisluku. Tällöin seuraavat väitteet pätevät.
\begin{enumerate}
    \item Jos $a \equiv b$ (mod $n$) ja $c \equiv d$ (mod $n$), niin $a + c \equiv b + d$ ja $ac \equiv bd$ (mod $n$)
    \item Jos $ca \equiv cb$ (mod $n$) ja $syt(c, n) = 1$, niin $a \equiv b$ (mod $n$)
    \item Jos $a \equiv b$ (mod $kn$) jollakin $k \in \Z$, niin $a \equiv b$ (mod $n$)
\end{enumerate}

Lause: Olkoot $a, c$ ja $n$ kokonaislukuja, ja olkoon $n$ lisäksi positiivinen. Kongruenssiyhtälöllä $ax = c$ (mod $n$) on kokonaislukuratkaisu, jos ja vain jos $syt(a, n)|c$

\section{Sykliset ryhmät}

\]

\end{example}

\]

Lause: Olkoon $n \in \Z$. Jäännösluokkien joukko $\Z_n$ on ryhmä, kun laskutoimituksena on yhteenlasku. Ryhmä $Z_n$ on syklinen, ja sen kertaluku on $n$.

\end{example}

\section{Ekvivalenssirelaatio}
Olio on relaatiossa toisen kanssa, jos ne toteuttavat tietyn suhdetta kuvaavan ehdon.

\subsection{Relaatio}
$R$ on joukon $A$ relaatio, $R \subset A \times A$, $aRb: (a, b) \in \R$

\begin{example}
$A = \{1,2,3,4,5\}, aRb, a + b < 6, (2,3) \in R, 2R3$
\end{example}

\subsection{Ekvivalenssirelaatio}
Määritelmä: Olkoon $R$ joukon $A$ relaatio. Kyseessä on ekvivalenssirelaatio, jos $R$ toteuttaa suraavat ehdot kaikilla $a, b, c \in A$.
\begin{enumerate}
    \item $aRa$ (refleksiivisyys)
    \item $aRb \to bRa$ (symmetrisyys)
    \item $aRb$ ja $bRc$ $\to$ $aRc$ (transitiivisuus)
\end{enumerate}

\subsection{Ekvivalenssiluokat}
$[a]_{\sim} = \{b \in A|b \sim a \}$

$a$ on ekvivalenssiluokan edustaja, $A / \sim$ kaikkien ekvivalenssiluokkien joukko

Lemma: Olkoon $\sim$ joukon $A$ ekvivalenssirelaatio. Jos $a \in [b]$ jollakin alkioilla $a, b \in A$, niin $[a] = [b]$

\subsection{Ositus}
Määritelmä: Olkoon $A$ jokin joukko. Kokoelmaa epätyhjiä $A$:n osajoukkoja nimitetään $A$:n ositukseksi, jos jokainen $a \in A$ kuuluu täsmälleen yhteen kokoelman osajoukkoon.

Lause: Olkoon $\sim$ joukon $A$ ekvivalenssirelaatio. Ekvivalenssiluokkien joukko $A/\sim$ on joukon $A$ ositus. 

Laiuse: Olkoon $I$ jokin indeksijoukko. Oletetaan, että joukot $C_i$, missä $i \in I$, muodostavat joukon $A$ osituksen. Määritellään relaatio $\sim$ seuraavasti. $a \sim b$, jos $a$ ja $b$ kuuluvat osajoukkoon $C_i$ jollakin $i \in I$. Tällöin $\sim$ on ekvivalenssirelaatio.
