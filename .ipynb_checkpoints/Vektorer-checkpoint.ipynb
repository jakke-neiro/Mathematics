\textbf{Samansuuntaiset, Lika riktade}

\textbf{Vastakkaisuuntaiset, Motsatt riktade}

\section{Vektorer i planet}
\subsection{Produkten av ett reelt tal och en vektorer}
\begin{enumerate}
    \item $|r\overline{a}| = |r||\overline{a}|$
    \item $r\overline{a} \uparrow \uparrow \overline{a}$, om $r > 0$
    \item $r \overline{a} \uparrow \downarrow \overline{a}$, om $r < 0$
\end{enumerate}

Delning av sträcka:
\[
\overline{v} = \frac{p \overline{a} + q \overline{b}}{p + q}
\]
Enhetsvektor: $\overline{a}^o = \frac{\overline{a}}{|\overline{a}|} = \frac{1}{|\overline{a}|}\overline{a}$

Samma riktning: $\overline{a}^o = \frac{1}{|\overline{a}|}\overline{a}$, $\frac{1}{|\overline{a}|} > 0$, $a^o \uparrow \uparrow \overline{a}$ \qed

Längd: $|a^o| = |\frac{1}{|\overline{a}|}\overline{a}| = |\frac{1}{|\overline{a}|}|\overline{a}| = 1$

\subsection{Parallella vektorer}
$\overline{a} \| \overline{b} \iff \overline{a} = r \overline{b}$

$\overline{a} \uparrow \uparrow \overline{b} \iff \overline{a} = r\overline{b}, r > 0$

$\overline{a} \uparrow \downarrow \overline{b} \iff \overline{a} = r \overline{b}, r < 0$

\begin{proof}
$\overline{a} \uparrow \uparrow \overline{b} \to \overline{a}^o = \overline{b}^o$

$\overline{a}^o = \frac{\overline{a}}{|\overline{a}}$, $\overline{a} = |\overline{a}|\overline{a}^o$, $\overline{a} = |\overline{a}|\overline{b}^o$, $\overline{a} = |\overline{a}| \frac{\overline{b}}{|\overline{b}|}$, $\overline{a} = \frac{|\overline{a}|}{|\overline{b}|}\overline{b}$, $\overline{a} = r \overline{b}$
\end{proof}

\subsection{Basvektorer}
Varke vektor $\overline{v}$ i planet kan uttryckas mha basvektorerna $\overline{a}$ och $\overline{b}$. En vektors uppdelning i komponenter är entydig i en bas.
\[
r \overline{a} + s \overline{b} = t \overline{a} + u \overline{b} \iff r = t, s= u
\]

\subsection{Skalär produkt}
\[
\overline{a} \cdot \overline{b} = |\overline{a}||\overline{b}|\cos \alpha
\]
var $\alpha$ är vinkeln mellan vektorerna.

\begin{enumerate}
    \item $\overline{a} \cdot \overline{a} = |\overline{a}|^2$
    \item $\overline{a} \cdot \overline{b} = \overline{b} \cdot \overline{a}$
    \item $\overline{a} \cdot (\overline{b} + \overline{c}) = \overline{a}\cdot\overline{b} + \overline{a}\cdot\overline{c}$
    \item $(s \overline{a}) \cdot (r \overline{b}) = (sr)(\overline{a} \cdot \overline{b})$
\end{enumerate}

$r\overline{i} + s \overline{j} = t \overline{i} + u \overline{j}$, $r = t$ och $s = u$

Ortsvektorn för punkten $P=(x,y)$ är $\overline{p} = \overline{OP} = x \overline{i} + y \overline{j}$

\textbf{Paikkavektori, Ortsvektor}

\subsection{Längden av en vektor}

\begin{example}
$\overline{a} = - 8\overline{i} - 6 \overline{j}$, $a =\sqrt{(-8)^2 + (-6)^2}, a = 10$
\end{example}

$= (x_2 - x_1)\overline{i} + (y_2 - y_1) \overline{j}$, $|\overline{AB}| = \sqrt{(x_2 - x_1)^2 + (y_2 - y_1)^2}$

\[
\overline{v} = r \overline{a} + s \overline{b} + t\overline{c}= u \overline{a} + v \overline{b} + w \overline{c} \iff r=u, s=v, t = w
\]

r \overline{i} + s \overline{j} + t \overline{k} = u \overline{i} + v \overline{j} + w \overline{k} \iff r=u, s=v, t=w

\[

\[
\begin{cases}

Allmän form

\frac{x - x_0}{s_x} = \frac{y - y_0}{s_y} = \frac{z - z_0}{s_z}
