\subsection{Singulaariarvohajotelma}
$A m \times n$ matriisi, aste $r > 0$. $\exists$ ortogonaalinen $m \times m$ matriisi $P$ ja ortogonaalinen $n \times n$ matriisi $Q$, siten että $A = PDQ'$ ja $D=P'AQ$, jossa $D m \times n$ on:
\begin{enumerate}
    \item $\Delta$, jos $r=m=n$
    \item $(\Delta O_{rx(n-r)}), r=m < n$
    \item $\begin{bmatrix}\Delta \\ O_{(m-n)\times n}\end{bmatrix}$, $r = n < m$
    \item $\begin{bmatrix}\Delta & O_{rx(n-r)} \\ O_{(m-r)\times r} & O_{(m-r) \times (n-r)} \end{bmatrix}$, $r < m$, $r < n$
\end{enumerate}

$\Delta$ on $r \times r$ diagonaalimatriisi, jonka diagonaalialkiot ovat positiivisia. Matriisin $\Delta^2$ diagonaalialkiot ovat matriisien $A'A$ ja $AA'$ positiiviset ominaisarvot. Matriisin $P$ sarakkeina ovat matriisin $AA'$ ominaisvektorit ja matriisin $Q$ sarakkeina matriisin $A'A$ ominaisvektorit.

Olkoon $A m \times n$ matriisi, jonka aste $r>0$. Tällöin $\exists m \times r$ matriisi $P$, ja $n \times r$ matriisi $Q$, siten että $P_1'P_1 = Q_1'Q_1 = I_r$ ja $A = P_1 \Lambda Q_1'$, jossa $A$ on $r \times r$ diagonaalimatriisi, diagonaalialkiot ovat positiivisia
\[
 P = \begin{bmatrix}P_1 & P_2 \end{bmatrix}
\]
\[
Q = \begin{bmatrix}Q_1 & Q_2 \end{bmatrix}
\]
Singulaariarvojen lukumäärä $r = A$:n aste

$P_1$ sarakkeet, $A$:n sarakeavaruuden ortonormaali kanta

$P_2$ sarakkeet, $A$:n riviavaruuden ortonormaali kanta

$P_2$ sarakkeet virittävät $A'$ nolla-avaruuden $N(A') = \{x: A'x = 0, x \in \R^m \}$

$Q_2$ sarakkeet virittävät $A$ nolla-avaruuden $N(A)$

Olkoon $A m \times m$ matriisi. Matriisin $A$ yleistetty $n \times m$ käänteismatriisi $A^{-}$ toteuttaa ehdon $AA^{-}A = A$

$A m \times m, r>0$ ja singulaariarvohajotelma

A = P\begin{bmatrix}\Delta & O_{rx(n-r)} \\ O_{(m-r)\times r} & O_{(m-r) \times (n-r)}\end{bmatrix}Q'

\[
B = Q \begin{bmatrix}\Delta^{-1} & E \\ F & G \end{bmatrix}P'

\subsection{Ositetun matriisin käänteismatriisi}
$m \times m$, $A = \begin{bmatrix}A_{11} & A_{12} \\ A_{21} & A_{22} \end{bmatrix}$, $A_{11} m_1 \times m_1$, $A_{12} m_1 \times m_2$, $A_{21} m_2 \times m_1$, $A_{22} m_2 \times m_2$

$A$ on ei-singulaarinen $m \times m$ matriisi, ositettu $B = A^{-1} = \begin{bmatrix}B_{11} & B_{12} \\ B_{21} & B_{22}\end{bmatrix}$ matriisin ositettu käänteismatriisi, jossa $B_{11}, N_{12}, B_{21}$ ja $B_{22}$ ovat samaa kokoa vastaavien matriisin $A$ alimatriisien kanssa.

Jos $A_{11}$ ja $A_{22} - A_{21}A_{11}^{-1}A_{12}$ ovat ei-singulaarisia, niin
\begin{enumerate}
    \item $B_{11} = A_{11}^{-1} + A_{11}^{-1}A_{12}B_{22}A_{21}A_{11}^{-1}$
    \item $B_{12} = -A_{11}^{-1}A_{12}B_{22}$
    \item $B_{22} = (A_{22} - A_{21}A_{11}^{-1}A_{12})^{-1}$
    \item $B_{21} = -B_{22}A_{21}A_{11}^{-1}$
\end{enumerate}

Jos matriisit $A_{22}$ ja $A_{11} - A_{12}A_{22}^{-1}A_{21}$ ovat ei-singulaarisia, niin
\begin{enumerate}
    \item $B_{11} = (A_{11} - A_{12}A_{22}^{-1}A_{21})^{-1}$
    \item $B_{12} = -B_{11}A_{12}A_{22}^{-1}$
    \item $B_{22} = A_{22}^{-1} + A_{22}^{-1}A_{21}B_{11}A_{12}A_{22}^{-1}$
    \item $B_{21} = -A_{22}^{-1}A_{21}B_{11}$
\end{enumerate}

\subsection{Ositetun matriisin determinantti}
$A$ on ei-singulaarinen $m \times m$ matriisi, ositettu. Jos $A_{12} = 0$ tai $A_{21} = 0$. niin $|A| = |A_{11}||A_{22}|$

$A$ on ei-singulaarinen $m \times m$ matriisi, ositettu
\begin{enumerate}
    \item $|A| = |A_{22}||A_{11} - A_{12}A_{22}^{-1}A_{21}|$, jos $A_{22}$ on ei-singulaarinen
    \item $|A| = |A_{11}||A_{22} - A_{21}A_{11}^{-1}A_{12}|$, jos $A_{11}$ on ei-singulaarinen
\end{enumerate}

$A m \times n$ matriisi ja $B n \times m$ matriisi. Tällöin
