\section{Discrete-time models}
\subsection{Density-dependent birth: site-based models}
$N_T$ parents produce $BN_T$ offspring and die
\begin{enumerate}
    \item The offspring are randomly and independently distributed over $M$ sites
    \item Within each site, the offspring compete and only one survives
\end{enumerate}

Skellam model: one offspring survives in each non-empty site
\[
N_{T+1} = M(1 - P(0)) = M \bigg(1 - e^{-\frac{B N_T}{M}} \bigg)
\]
\[
x_T = \frac{N_T}{M}, x_{T+1} = 1 - e^{-B x_T}
\]


\section{Evolution of dispersal}
Annual plant in $M \tp \infty$ living sites

$B \to \infty$ seeds per plant, a fraction $x$ of the seeds disperse ($x=$ dispersal strategy)

Dispersed seeds survive (land in a site) with probability $p$

In each site, a randomly chosen seed becomes an adult plant (all others perish)

Kin competition (Hamilton and May 1977)

Currently present: $x_1, \dots, x_k$, $N_t^{(1)}, \dots, N_t^{(k)}$, $\sum_i N_t^{(i)} = M$. Skellam model with $B \to \infty$.

\[
N_{t+1}^{(i)} = N_t^{(i)}\bigg(P(\textrm{retaining home site}) + B x_i p P(\textrm{winning a site after dispersal}) \bigg)
\]

\[
P(\textrm{retaining home site}) = 
\frac{B(1-x_i)}{B(1-x_i)+ \sum_j N_t^{(j)}B x_j p/M}
\]

\[
P(\textrm{winning a site after dispersal}) = 
\sum_l \frac{N_t^{(l)}}{M}\frac{1}{1 + B(1-x_l) + \sum_j N_t^{(j)}B x_j p/M}
\]

\[
N_{t+1}^{(i)} = N_t^{(i)}\bigg(P(\textrm{retaining home site}) + B x_i p P(\textrm{winning a site after dispersal}) \bigg)
\]

\[
n_t^{(i)} = \frac{N_t^{(i)}}{M}
\]

\[
n_{(t+1)}^{(i)} = n_t^{(i)}\bigg(\frac{1-x_i}{1 - x_i + \sum_j n_t^{(j)}x_j p} + x_i p \sum_l n_t^{(l)}\frac{1}{1 - x_l + \sum_j n_t^{(j)}x_j p} \bigg)
\]

\[
n_{t+1}^{(i)} = n_t^{(i)}\lambda (x_i, E_{1,t}, E_{2,t})
\]

Equilibrium: $\lambda(x_i, \hat{E}_1, \hat{E}_2) =1$ 

Invasion analysis (2 strains), Coexistence by mutual invasibility, globally stable attractors when restricted to the boundary but transversally both unstable

Invasion fitness of $x_2$ as a rare strain in the resident population of $x_1$
\[
W_{x_1}(x_2) = \lim_{n_2 \to 0, n_1 \to 1} \lambda(x_2, E_1, E_2)
\]
\[
\lambda(x, E_1, E_2) = \frac{1-x}{1-x+E_1} + x p E_2
