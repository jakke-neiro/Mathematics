\section{Reaaliluvut ja funktion määritelmä}
\subsection{Reaalilukujen kunta- ja järjestysaksioomat}
\begin{enumerate}
    \item $x + y = y + x \forall x, y \in \R$ (vaihdannaisuus)
    \item $(x+y) + z = x + (y + z) \forall x, y, z \in \R$ (yhteenlaskun liitännäisyys)
    \item $x + 0 = x$ (nollaehto)
    \item $x + y = 0, y = -x$ (Vastaluku)
    \item $xy = yx$ (vaihdannaisuus)
    \item $(xy)z = x(yz) \forall x, y, z$
    \item $x(y + z) = xy + xz$
    \item $1x = x$
    \item $xy = 1, y = 1/x$ Käänteisluku
\end{enumerate}

Lause: Reaalilukujen joukossa on yksi nolla-alkio ja yksi ykkösalkio.

Lause: Olkoot $x, y, z \in \R$. Jos $x+y = x + z$, niin $y = z$.

Lause: $a, b \in \R$. Yhtälöllä $a + x = b$ on yksikäsitteinen ratkaisu $x \in \R$, jota merkitään $b-a$ ja kutsutaan lukujen $b$ ja $a$ erotukseksi. Vastaavasti, jos $a \neq 0$, niin yhtälöllä $ax = b$ on yksikäsitteinen ratkaisu $x \in \R$, jota merkitään $b/a$, osamäärä.

Järjestysaksioomat:
\begin{enumerate}
    \item $x, y \in \R$ yksi on voimassa $x < y$, $x = y$, $x > y$
    \item $x, y, z \in \R$, $x > y \iff y < x$
    \item $x, y, z$ Jos $x < y$ ja $y < z$, niin $x < z$
    \item $x, y \in \R$ $x>0$ ja $y > 0$, niin $xy>0$
    \item $x, y, z \in \R$ $x < y$ niin $x+z < y +z$
\end{enumerate}

Lause: $x, y \in \R$. Tällöin $x=y$, jos ja vain jos $x \leq y$ ja $x \geq y$.

Aksioomajärjestelmää on täydennättävä täydellisyysaksioomalla.

\subsection{Infimum, supremum ja täydellisyysaksiooma}
Määritelmä: Olkoon $E$ reaalilukujen $\R$ epätyhjä osajoukko. Jos on olemassa sellainen $s \in \R$, että $x \leq s$ kaikilla $x \in E$, niin lukua $s$ sanotaan joukon ylärajaksi ja joukkoa $E$ sanotaan ylhäältä rajoitetuksi. Vastaavasti, jos on olemassa sellainen $p \in \R$, että $x \geq p$ kaikilla $x \in E$, niin lukua $p$ sanotaan joukon $E$ alarajaksi ja joukkoa $E$ sanotaan alhaalta rajoitetuksi. 

Määritelmä: Jos $E$ reaalilukujen epätyhjä ylhäältä rajoitettu osajoukko ja jos $s$ on joukon $E$ yläraja, joka toteuttaa epäyhtälön $s \leq x$ kaikilla joukon ylärajoilla $x$, niin $s$ on joukon $E$ supremum eli pienin yläraja. Tällöin $s = \sup E$.

Määritelmä: Jos $E$ on reaalilukujen epätyhjä alhaalta rajoitettu osajoukko ja jos $p$ on joukon $E$ alaraja, joka toteuttaa epäyhtälön $p \geq x$ kaikilla joukon alarajoilla $x$, niin $p$ on joukon $E$ infimum eli suurin alaraja. Tällöin $p = \inf E$. 

Lause:
\begin{enumerate}
    \item Jos joukossa $E$ on suurin luku $s$, niin $s = \sup E$.
    \item Jos joukossa $E$ on pienin luku $p$, niin $p = \inf E$
\end{enumerate}

Täydellisyysaksiooma: Ylhäältä rajoitetulla epätyhjällä reaalilukujen osajoukolla on aina supremum/infimum.

Lause: Supremum ja infimum ovat yksikäsitteisiä.

Lause: $E$ on reaalilukujen $\R$ epätyhjä ylhäältä rajoitettu osajoukko. Jokaista $\epsilon > 0$ kohti on olemassa $x \in E$, jolle 
\[
\sup E - \epsilon < x \leq \sup E
\]
\[
\inf E \leq x < \inf E + \epsilon
\]


\section{Lukujonon raja-arvo}
\subsection{Itseisarvo}
\textbf{Itseisarvo, Absolutbelopp}

Reaaliluvun $x$ itseisarvo $|x|=\begin{cases}x & x \geq 0 \\ -x & x < 0\end{cases}$

Kolmioepäyhtälö: $||x| + |y|| \leq |x+y| \leq |x| + |y|$

$||x| - |y|| \leq |x - y| \leq |x| + |y|$

Epäyhtälö $|x| \leq |y|$ on voimassa, jos ja vain jos $x^2 \leq y^2$

\subsection{Lukujonon raja-arvo}
Määritelmä: Jos jokaista lukua $n \in \N_0$ tai $n \in \N-1$ asetetaan vastaamaan reaaliluku $x_n$, niin syntyy lukujono $x_0, x_1, \dots, x_n$ tai $(x_n)$.

Lukujono on siis funktio $N_0 \to \R$ tai funktio $N_1 \to \R$.

Määritelmä: Lukujono $x_n$ suppenee kohti lukua $a \in \R$, jos jokaista lukua $\epsilon > 0$ vastaa sellainen luku $n_{\epsilon} \in \N_0$, että
\[
|x_n - a| < \epsilon \textrm{ kun } n > n_{\epsilon}
\]
Tällöin sanotaan, että lukujonolla $x_n$ on raja-arvo $a$ ja merkitään $\lim_{n \to \infty}x_n = a$ tao $x_n \to a$, kun $n \to \infty$.

\begin{example}
$x_n = (-1)^n \frac{s}{\frac{1}{2}n + 1} \forall n \in N_0$, suppenee kohti $N_0$

$|x_n - 0| = |(-1)^n \frac{5}{\frac{1}{2}n + 1} - 0| = \frac{5}{\frac{1}{2}n + 1} = \frac{10}{n+2} < \frac{10}{n}$

$\frac{10}{n} <  \epsilon$, $n_{\epsilon} \geq \frac{10}{\epsilon}$

Ratkaisu: Olkoon $\epsilon > 0$, $n_{\epsilon} \geq \frac{10}{\epsilon}$

$\bigg|(-1)^n \frac{5}{\frac{1}{2}n + 1} - 0 \bigg| = \frac{10}{n+2} < \frac{10}{n} < \frac{10}{10/\epsilon} = \epsilon, n > n_{\epsilon}$
\end{example}

Lause: $x_n$ suppeneva lukujono. Jos $\lim x_n = a$ ja $\lim x_n = b$, niin $a = b$, toisin sanoen lukujonon raja-arvo on yksikäsitteinen.

Lause: $x_n$ suppeneva lukujono. Tällöin on olemassa sellainen $M > 0$, että $\{x_n: n \in N_1\} \subset [-M, M]$

Lause: Olkoon $x_n$ lukujono, joka suppenee kohti lukua $x$. Olkoon $y_n$ lukujono, joka suppenee kohti lukua $y$. Tällöin
\begin{enumerate}
    \item Lukujono $(x_n+y_n)$ suppenee ja $\lim_{n \to \infty}(x_n + y_n) = x + y$
    \item $\forall r \in \R$ lukujono $rx_n$ suppenee ja $\lim_{n \to \infty}rx_n = rx$
    \item Lukujono $x_n y_n$ suppenee ja $\lim_{n \to \infty}(x_n y_n) = xy$
    \item Lukujono $x_n/y_n$ suppenee ja $\lim_{n \to \infty}\frac{x_n}{y_n} = \frac{x}{y}$
\end{enumerate}

Lause: Olkoon $x_n$ lukujono, joka suppenee kohti lukua $x$, ja olkoon $y_n$ lukujono, joka suppenee kohti lukua $y$.
\begin{enumerate}
    \item Jos $x_n \leq y_n$ $\forall n$, niin $x \leq y$
    \item Jos $x_n \geq y_n$ $\forall n$, niin $x \geq y$
\end{enumerate}

Lause: Olkoon $x_n$ lukujono, joka suppenee kohti lukua $x$. Tällöin sen jokainen osajono suppenee kohti lukua $x$.

\subsection{Monotoninen lukujono ja rajatta kasvaminen}
Määritelmä: Lukujono $x_n$ on kasvava, jos $x_n \leq x_{n+1} \forall n$. Lukujono on aidosti kasvava, jos $x_n < x_{n+1} \forall n$. Lukujono $x_n$ on vähenevä, jos $x_n \geq x_{n+1} \forall n$. Lukujono on aidosti vähenevä, jos $x_n > x_{n+1} \forall n.$ Lukujono on monotoninen, jos se on kasvava tai vähenevä, aidosti monotoninen jos aidosti kasvava tai vähenevä.

Lause: $x_n$ kasvava lukujono. Jos $\exists M \in \R$, jolle $x_n \leq M \forall n$, niin lukujono suppenee. 
\[
\lim_{n \to \infty}x_n = \sup\{x_n: n \in \N_1 \} \leq M
\]

Lause: $x_n$ vähenevä lukujono. Jos $\exists m \in \R$, jolle $x_n \geq m \forall n$, niin lukujono suppenee. Tällöin 
\[
\lim_{n \to \infty}x_n = \inf\{x_n : n \in N_1 \} \geq m
\]

\begin{example}
$x_n = \frac{n}{n+1}$, $\forall n \in N_1$
\[
\frac{x_n + 1}{x_n} = \frac{\frac{n+1}{n+1+1}}{\frac{n}{n+1}} = \frac{(n+1)^2}{n(n+2)} = \frac{n^2 + 2n +1}{n^2 + 2n} > 1
\]
$x_n < x_n + 1$ aidosti kasvava

$0 < x_n < 1 \to$ lukujono suppenee
\[
\sup\{\frac{n}{n+1}: n \in \N\} = 1 \frac{n}{n+1} > \frac{m}{m+1} = \frac{m+1-1}{m+1} = 1 - \frac{1}{m+1} > 1 - \frac{1}{m}
\]
\end{example}

Lause: Bolzanon-Weierstrassin lause, Olkoon $x_n$ rajoitettu lukujono eli lukujono, jonka jäsenet kuuluvat rajoitetulle välille. Tällöin lukujonolla on suppeneva osajono. 

\begin{example}
$x_n = \bigg(1 + \frac{1}{n} \bigg)^n$ aidosti kasvava ja ylhäältä rajoitettu?
\[
\frac{x_n+1}{x_n} = \bigg(1 - \frac{1}{(n+1)^2}\bigg)^{n+1} \frac{n+1}{n} > 1
\]
Bernoullin epäyhtälö
\[
\bigg(1 + \frac{1}{n} \bigg)^n < \bigg(1 - \frac{1}{n} \bigg)^{-n} \leq \bigg(1 - \frac{1}{2} \bigg)^{-2} = 4
\]
\[
x_n = 1 + \sum_{k=1}^n {n \choose k} \frac{1}{n^k} \leq 1 + \sum_{k=1}^n \frac{1}{2^{k-1}} = 1 + \frac{1 - (1/2)^n}{1-1/2} = 3
\]
\end{example}

Määritelmä: Lukujonon $((1 + \frac{1}{n})^n)$ raja-arvo on Neperin luku $\epsilon$.

Määritelmä: Lukujono $(x_n)$ kasvaa rajatta, jos jokaista reaalilukua $m$ kohti on olemassa sellainen$n_{\epsilon} \in \N_0$, että
\[
x_n > m \forall n > n_m \lim_{n \to \infty}x_n = \infty
\]
Lukujono $(x_n)$ vähenee rajatta, jos jokaista reaalilukua $m$ kohti on olemassa sellainen $n_m \in \N_0$, että
\[
x_n < m \forall n > n_m \lim x_n = - \infty
\]

\subsection{Cauchyn yleinen suppenemisehto}
Lause: Olkoon $x_n$ suppeneva lukujono. Tällöin jokaista lukua $\epsilon > 0$ vastaa sellainen $n_{\epsilon} \in \N_0$, että kun $n>n_{\epsilon}$, niin
\[
|x_n - x_{n+k}| < \epsilon \forall k \in N_0
\]

Lemma: Olkoon $[a_n, b_n], n \in \N$ välejä, joille $[a_{n+1}, b_{n+1}] \subset [a_n, b_n] \forall n \in \N$ ja
\[
\lim_{n \to \infty}(b_n - a_n) = 0
\]
Tällöin $\bigcap_{n=1}^{\infty}[a_n, b_n]$ sisältää yhden luvun. 

Lause: Lukujono $x_n$ suppenee, jos ja vin jos jokaista lukua $\epsilon > 0$ kohti on olemassa luku $n_{\epsilon} \in \N_0$ siten, että
\[
|x_n - x_{n+k}| < \epsilon \forall n > n_{\epsilon} \forall k \in N_0
\]

\section{Funktion raja-arvo}
\subsection{Raja-arvon määritelmä ja perusominaisuudet}
Määritelmä: Olkoot $f: A \to \R$ ja $x_0 \in \R$. Oletetaan, että on olemassa sellainen $r>0$, että $(x_0 - r,x_0) \cup (x_0, x_0 + r) \subset A$. Tällöin funktiolla $f$ on pisteessä $x_0$ raja-arvo $m \in \R$, jos jokaista $\epsilon > 0$ kohti on olemassa sellainen $\delta > 0$, että
\[
|f(x) - m| < \epsilon, \textrm{ kun }0 < |x - x_0 | < \delta
\]
\[
\lim_{x \to x_0}f(x) = m
\]

\begin{example}
$f: \R \to \R$, $f(x) = 2x - 1$. Osoita $\lim_{x \to 1}f(x) = 3$

$\epsilon > 0$, $\delta = \frac{1}{2}\epsilon$, $0 < |x-1| < \delta$

$|2x +1 - 3| = |2x-2| = 2|x-1| < 2\delta = 2 \frac{1}{2} \epsilon = \epsilon$
\end{example}

\begin{example}
$f: \R \to \R$, $f(x) = x^3$ pisteessä $2$ raja-arvo $8$

$|f(x) - 8| = |x^3 - 8| = |x-2||x^2 + 2x +4|$ rajoittamalla $\delta < 1$, $0 < |x - 2| < \delta < 1$

$|x-2| < 1 \iff x \in (1,3)$, $|x^2 + 2x + 4| \leq x^2 + 2x + 4 = 19$

$|f(x) - 8| < 19|x-2| \to |x-2| < 1, |x-2| < \epsilon/19$

Ratkaisu: Olkoon $\epsilon > 0$. Valitaan $\elta = \min\{1, \frac{\epsilon}{19} \}$. Saamme
\[
|f(x) - 8| = |x^3 - 8| = |x-2||x^2 + 2x +4| < 19|x-2| < 19 \delta = 19 \frac{\epsilon}{19} = \epsilon
\]
\end{example}

\begin{example}
Olkoon $f:(0,1) \to \R$, $f(x) = \frac{7}{x}$, $\lim_{x \to 1/2}f(x) = 14$

$|f(x) - 14| = |\frac{7}{x} - 14| = \bigg| \frac{7 -14x}{x} \bigg| = \frac{12|1/2 - x|}{|x|}$

Raja-arvo kohdassa $1/2$, oletamme $x > 1/4$ ja $\frac{14}{|x|} < 4\cdot14 = 56$

Ratkaisu. Olkoon $\epsilon > 0$. Valitsemme $\delta = \textrm{min}\{1/4, \epsilon/56 \}$. Lasketaan
\[
|f(x) - 14| = |7/x - 14| = \bigg|\frac{7 - 14x}{x}\bigg| = \frac{14|1/2 - x|}{|x|}
\]
Jos $|1/2 - x| < \delta \leq 1/4$, niin $x \geq 1/4$
\[
|f(x) - 14| \leq \frac{14|1/2 - x|}{1/4} = 56|1/2 - x| < 56\delta \leq \epsilon
\]
\end{example}

Lause: Olkoot $f: A \to \R$ ja $x_0 \in \R$. Jos $\lim_{x \to x_0}f(x)=m_1 \in \R$ ja $\lim_{x \to x_0}f(x) = m_2 \in \R$, niin $m_1 = m_2$ eli funktion raja-arvo on yksikäsitteinen. 

Lause: Olkoot $f: A \to \R$ ja $x_0 \in \R$. Jos funktiolla $f$ on pisteessä $x_0$ raja-arvo, niin jokaista $\epsilon > 0$ kohti on olemassa sellainen $\delta > 0$, että $|f(x) - f(y)| < \delta$, kaikilla luvuilla $x, y \in A$, joilla $0 < |x - x_0| < \delta$ ja $0 < |y - x_0| < \delta$.

Lause: Olkoot $f: A \to \R$ ja $x_0 \in \R$. Oletetaan, että on olemassa sellainen $r>0$, että $(x_0-r, x_0) \cup (x_0, x_0 + r) \subset A$. Seuraavat ehdot ovat yhtäpitäviä: 
\begin{enumerate}
    \item Funktiolla $f$ on pisteessä $x_0$ raja-arvo $m$
    \item Olkoon $(y_n)$ lukujono, jolle $y_n \neq x_0$ kaikilla $n \in \N_1$. Jos lukujono $(y_n)$ suppenee kohti lukua $x_0$, niin lukujono $(f(y_n))$ suppenee kohti lukua $m$
\end{enumerate}

Lause: Olkoon $f: A \to \R$ funktio, jolla on raja-arvo pisteessä $x_0 \in \R$. Tällöin on olemassa sellaiset $m, \elta > 0$, että $|f(x)| < m$ kun $0 < |x - x_0| < \delta$.

Lause: Olkoon $f: A \to \R$ funktio, jolla on raja-arvo pisteessä $x_0 \in \R$.
\begin{enumerate}
    \item Jos on olemassa sellainen $M \in \R$, että $f(x) \leq M$ kaikilla $x \in A$, niin $\lim_{x \to x_0}f(x) \leq M$
    \item Jos on olemassa sellainen $m \in \R$, että $f(x) \geq m$ kaikilla $x \in A$, niin $\lim_{x \to x_0}f(x) \geq m$
\end{enumerate}

Lause: Olkoot $f: A \to \R$ ja $g: A \to \R$ funktioita, joille $\lim_{x \to x_0}f(x) = a$ ja $\lim_{x \to x_0}g(x) = b$. Olkoon $m \in \R$. Tällöin
\begin{enumerate}
    \item Funktiolla $f + g: A \to \R$ on raja-arvo pisteessä $x_0$ ja $\lim_{x \to x_0}(f(x) + g(x)) = a + b$.
    \item Funktiolla $mf:A \to R$ on raja-arvo pisteessä $x_0$ ja $\lim_{x \to x_0}mf(x) = ma$
    \item Funktiolla $fg: A \to \R$ on raja-arvo pisteessä $x_0$ ja $\lim_{x \to x_0}f(x)g(x) = ab$
    \item Jos lisäksi $b \neq 0$, niin funktiolla $f/g: A \subset \{x: g(x) = 0 \} \to \R$ on raja-arvo pisteessä $x_0$ ja $\lim_{x \to x_0}\frac{f(x)}{g(x)} = a/b$
\end{enumerate}

Määritelmä: Olkoot $f: A \to \R$ ja $x_0 \in \R$. Oletetaan, että on olemassa sellainen $r > 0$, että $(x_0 - r, x_0) \cup (x_0, x_0 + r) \subset A$. Funktiolla $f$ on pisteessä $x_0$ raja-arvo $\infty$, jos jokaista $m \in \R$ kohti on olemassa sellainen $\delta > 0$, että $f(x) > m$, kaikilla $x \in A$, joilla $0 < |x - x_0| < \delta$. $\lim_{x \to x_0}f(x) = \infty$ tai $f(x) \to \infty$ kun $x \to x_0$. 

Funktiolla $f$ on pisteessä $x_0$ raja-arvo $-\infty$, jos jokaista $m \in \R$ kohti on olemassa sellainen $\delta > 0$, että $f(x) < m$ kaikilla $x \in A$, $0 < |x-x_0| < \delta$. Tällöin merkitään $\lim_{x \to x_0}f(x) = -\infty$

\begin{example}
$f: (-\infty, 0) \cup (0, \infty) \to \R, f(x) = \frac{1}{|x|}$, $\lim_{x \to 0}f(x) = \infty$

Olkoon $m \in \R$. Jos $m \leq 0$, niin $f(x) > m \forall x \in \R \setminus\{0\}, m>0$

$\delta = \frac{1}{m}$, $f(x) = \frac{1}{|x|} > \frac{1}{\delta} = m$, kun $0 < |x| < \delta$
\end{example}

\subsection{Toispuoleiset raja-arvot ja raja-arvo äärettömässä}
Määritelmä: Olkoot $f: A \to \R$ ja $x_0 \in \R$. Oletetaan, että on olemassa sellainen $r > 0$, että $(x_0 - r, x_0) \subset A$. Tällöin funktiolla $f$ on pisteessä $x_0$ vasemmanpuoleinen raja-arvo $m$, jos jokaista $\epsilon > 0$ kohti on olemassa sellainen $\delta > 0$, että 
\[
|f(x) - m| < \epsilon \forall x \in A
\]
joilla $x_0 - \delta < x < x_0$. $\lim_{x \to x_0^-}f(x) = m$

Oletetaan, että on olemassa sellainen $r>0$, että $(x_0, x_0 + r) \subset A$. Tällöin funktiolla $f$ on pisteessä $x_0$ oikeanpuoleinen raja-arvo $m$, jos jokaista $\epsilon > 0$ kohti on olemassa sellainen $\delta > 0$, että $|f(x) - m| < \epsilon \forall x \in A$, joilla $x_0 < x < x_0 + \delta$. $\lim_{x \to x_0^+}f(x) = m$.

Lause: Olkoot $f: A \to \R$ ja $x_0 \in \R$. Oletetaan, että on olemassa $r>0$ siten, että $(x_0-r, x_0) \cup (x_0, x_0 + r) \subset A$. Tällöin funktiolla $f$ on pisteessä $x_0$ raja-arvo $m$, $\lim_{x \to x_0}f(x) = m$, jos ja vain jos funktiolla $f$ on pisteessä $x_0$ sekä vasemman- että oikeanpuoleinen raja-arvo $m$, $\lim_{x \to x_0^-}f(x) = \lim_{x \to x_0^+}f(x) = m$.

Määritelmä: Olkoon $f: A \to \R$. Oletetaan, että on olemassa sellainen $r \in \R$, että $(r, \infty) \subset A$. Tällöin funktiolla $f$ on positiivisessa äärettömyydessä raja-arvo $m$, jos jokaista $\epsilon > 0$ kohti on olemassa sellainen $h \in \R$, että 
\[
|f(x) - m| < \epsilon \forall x > h \lim_{x \to \infty}f(x) = m \forall x < h \lim_{x \to -\infty}f(x) = m
\]
\begin{example}
$f:(1, \infty) \to \R$, $f(x) = 1/x$

$\epsilon > 0, \delta = \textrm{max}\{1, 1/\epsilon\}$, $|f(x) - 0| = 1/x < \epsilon$, $x > \delta$, $\lim_{x \to \infty}f(x) = 0$
\end{example}

\subsection{Monotoninen funktio}
Määritelmä: Funktio $f: A \to \R$ on
\begin{enumerate}
    \item Kasvava, jos $f(x_1) \leq f(x_2) \forall x_1, x_2 \in A$, $x_1 \leq x_2$
    \item Aidosti kasvava, jos $f(x_1) < f(x_2) \forall x_1, x_2 \in A$, $x_1 < x_2$
    \item Vähenevä, jos $f(x_1) \geq f(x_2) \forall x_1, x_2 \in A$, $x_1 \geq x_2$
    \item Aidosti vähenevä, jos $f(x_1) > f(x_2) \forall x_1, x_2 \in A$, $x_1 > x_2$
    \item Monotoninen
    \item Aidosti monotoninen
\end{enumerate}

\begin{example}
$f_n:[0, \infty) \to \R$, $f_n(x) = x^n$ ovat aidosti kasvavia kaikilla $n \in N_1$

$f_{n+1}(x_1) = x_1^{n+1} = x_1x_1^n < x_1x_2^n < x_2x_2^n = x_2^{n+1} = f_{n+1}(x_2)$
\end{example}

Lause: Olkoon $f:(a, b) \to \R$ monotoninen funktio.
\begin{example}
    \item Jos joukko $f((a, b))$ on alhaalta rajoitettu, niin funktiolla $f$ on pisteessä $a$ oikeanpuoleinen raja-arvo
    \item Jos joukko $f((a, b))$ on ylhäältä rajoitettu, niin funktiolla $f$ on pisteessä $b$ vasemmanpuoleinne raja-arvo
\end{example}

Lause: Jos $f:[a, b] \to \R$ on monotoninen, niin sillä on oikean- ja vasemmanpuoleiset raja-arvot jokaisessa pisteessä $x_0 \in (a, b)$. Lisäksi, jos $f$ on kasvava, niin
\[
\lim_{x \to x_0^-}f(x) \leq f(x_0) \leq \lim_{x \to x_0^+}f(x)
\]
kaikilla $x_0 \in (a, b)$ sekä $f(a) \leq \lim_{x \to a^+}f(x)$ ja $\lim_{x \to b^-}f(x) \leq f(b)$

Vastaavasti, jos $f$ on vähenevä, niin
\[
\lim_{x \to x_0^-}f(x) \geq f(x_0) \geq \lim_{x \to x_0^+}f(x)
\]
kaikilla $x_0 \in (a, b)$ sekä $f(a) \geq \lim_{x \to a^+}f(x)$ ja $\lim_{x \to b^-}f(x) \geq f(b)$
