\section{Johdanto ja terminologiaa}
Tavallinen differentiaaliyhtälö: $y''(x) + (y'(x))^2 = \sin(x + y(x))$

Määritelmä: Olkoon $F:D \to \R$ jatkuva funktio. $D \subset \R$ on avoin ja yhtenäinen.
\[
F(x, y(x), y'(x), \dots, y^{(n)}(x)) = 0
\]
Normaalimuotoinen yhtälö: $y^{(n)}(x) = f(x, y(x), \dots, y^{(n-1)}(x))$

Määritelmä: Differentiaaliyhtälön lokaalilla ratkaisulla tarkoitetaan avoimella välillä $I$ määriteltyä $n$-kertaa derivoituvaa funktiota $y:]a, b[ \to \R$, jolle
\[
(x, y(x), \dots, y^{n}(x)) \in D \forall x \in ]a, b[
\]
ja 
\[
F(x, y(x), y'(x), \dots, y^{(n)}(x)) = 0 \forall x \in ]a, b[
\]

Primitiivinen funktio:

$y' = f(x)$, $y(x) = \int^x f(t) dt + C$, $C \in \R$

Alkuarvo $y(x_0) = y_0$, $y(x) = \int_{x_0}^x f(t)dt + y_0$

\subsection{Esitietoja}
Lause(Analyysin peruslause I): Olkpoon $f:[a,b] \to \R$ jatkuva ja $F:[a,b] \to \R$.
\[
F(x) = \int_c^x f(t)dt, c \in [a,b]
\]
Tällöin $F$ on derivoituva välillä $[a,b]$ ja $F'(x) = f(x)$, $\forall x \in [a, b]$

Jos $F:[a,b] \to \R$ on jatkuvan funktion $f: [a, b] \to \R$ primitiivi, eli derivoituva funktio, jolle $F'(x) = f(x)$, $\forall x \in [a, b]$
\[
F(x) = \int_c^xf(t)dt + F(c), c \in [a, b]
\]
\[
F(x) = \int^x f(t) dt
\]

Lause (Analyysin peruslause 2): Olkoon $f: \R \to \R$ jatkuvasti derivoituva. Tällöin
\[
f(x) - f(x_0) = \int_{x_0}^x f'(t) dt \forall x, x_0 \in \R
\]

\section{Ensimmäisen kertaluvun yhtälöt}
$F(x, y(x), y'(x)) = 0$, $y'(x) = f(x, y(x))$

$D = ]a, b[ \times ]c, d[$ $(a > b, c <d)$, $f$ on jatkuva, ratkaisut jatkuvasti derivoituvia

\subsection{Separoituvat yhtälöt}
Jatkuvat funktiot $g:]a, b[ \to \R$, $h:]c, d[ \to \R$, $y'(x) = g(x)h(y(x))$

Lause: Olkoon $y'=g(x)h(y)$ separoituva yhtälö ja $h(y) \neq 0 \forall y$. 
\[
\int^y\frac{1}{h(t)}dt = \int^x g(s)ds + C, C \in \R
\]
Alkuarvon $y(x_0) = y_0$ toteuttava yksikäsitteinen ratkaisu
\[
\int_{y_0}^{y}\frac{1}{h(t)}dt = \int_{x_0}^x g(s) ds
\]

\begin{example}
\[
y' = \frac{(1+y^2)\sin x}{y}
\]
\[
\int^y\frac{t}{1+t^2}dt = -\int^x \sin(s)ds + C, C \in \R
\]
\[
\frac{1}{2}\log(1 + y^2) = \cos x + C, C \in \R
\]
\[
y = \pm \sqrt{De^{2\cos x - 1}}, D = e^{2C} > 0
\]
\end{example}

\subsubsection{Lineaariset yhtälöt}
Määritelmä: $y'(x) + p(x)y(x) = q(x)$, $p, q: ]a, b[$ ovat jatkuvia. $q = 0 \to$ homogeeninen, $p=c, c \in \R \to$ vakiokertoiminen
\[
y(x) = e^{-\int^x p(t) dt} \bigg(\int^x e^{\int^t p(s) ds}q(t) + C \bigg), C \in \R 
\]

\begin{example}
$y' + y = x^2$

$y = e^{\int^x 1 dt} \bigg(\int^x e^{\int^t 1ds} t^2 dt + C \bigg)$

$= e^{-x} \bigg( \int^x e^t t^2 dt + C \bigg)$, $C \in \R$
\end{example}

Lause: $y' + ay = q(x)$
\begin{enumerate}
    \item Jos $q(x)$ on polynomi korkeintaan astetta $n$, niin yhtälöllä on korkeintaan astetta $n$ oleva polynomiratkaisu
    \item Jos $q(x) = Ae^{bx}$, niin ratkaisu on muotoa
    \[
    y (x) = \begin{cases}
    Ke^{bx} & b \neq -a, K \in \R\\
    Kxe^{bx} & b = -a, K \in \R \\
    \end{cases}
    \]
    \item Jos $q(x) = A \cos (\omega x) + B \sin (\omega x)$, $\omega \neq 0$. $y(x) = K \cos (\omega x) + L \sin (\omega x)$, $K, L \in \R$
\end{enumerate}

\subsection{Eksaktit yhtälöt}
Määritelmä: Olkoot $P$ ja $Q$ jatkuvia funktioita. Differentiaaliyhtälö $Q(x, y)y' + P(x, y) = 0$ on eksakti, jos on olemassa jatkuvasti differentioituva $u$ siten, että $\frac{\partial u}{\partial z}(x, y) = P(x, y)$ ja $\frac{\partial u}{\partial y}(x, y) = Q(x, y)$.

Lause: Olkoot $P, Q : D \to \R$, missä $D = ]a,b[ \times ]c, d[$, jaktuvasti differentioituvia. Tällöin $Q(x, y)y' + P(x, y) = 0$ on eksakti silloin kun $\frac{\partial P}{\partial y}(x, y) = \frac{\partial Q}{\partial x}(x, y)$, $\forall (x, y) \in D$. 

\begin{example}
$(3x + \cos y)y' + 3y + e^x = 0$

Eksakti: $\frac{\partial u}{\partial y} = (3y + e^x) = 3 = \frac{\partial}{\partial x}(3x + \cos y) = 3$

$\frac{\partial u}{\partial x} = 3y + e^x$, $u(x, y) = 3yx + e^x + g(y)$

$\frac{\partial u}{\partial y} = 3x + \cos y = 3x + g'(y)$, $g'(y) = \cos y$, $g(y) = \sin y$

Implisiittinen muoto: $3yx + e^x + \sin y = u(x, y) = C, C \in \R$
\end{example}

\subsubsection{Integroiva tekijä}
Palautetaan epäeksakti lause eksaktiksi!

Lause: Jos funktio $\gamma(y) = \frac{1}{P(x, y)} \bigg( \frac{\partial P}{\partial y} - \frac{\partial Q}{\partial x} \bigg)$ on muuttujan $x$ suhteen vakiofunktio niin yhälöstä $Q(x, y)y' + P(x, y) = 0$ saadaam eksakti kertomalla sitä puolittain funktiolla $\mu(y) = e^{-\int^y \gamma(t) dt}$

Lause: Jos funktio $\psi(x) = \frac{1}{Q(x, y)} \bigg( \frac{\partial P}{\partial y} - \frac{\partial Q}{\partial x} \bigg)$ on muuttujan $y$ suhteen vakiofunktio niin yhtälöstä $Q(x, y)y' + P(x, y) = 0$ saadaan eksakti kertomalla sitä puolittain funktiolla $\mu(x) = e^{\int^x \psi(t) dt}$

\section{Toisen kertaluvun yhtälöt}
\subsection{Olemassaolo- ja yksikäsitteisyys}
\[
y''(x) = f(x, y(x), y'(x)), f: D \to \R, D \subset \R^3
\]
\[
\begin{cases}
y' = z \\
z' = f(x, y, z) \\
\end{cases}
\]

Lause: Oletetaan, että $f: D \to \R$ ja sen osittaisderivaatat $\frac{\partial f}{\partial y}$, $\frac{\partial f}{\partial y'}$ ovat jaktuvia alueessa $D \subset \R^3 ja (x_0, y_0, y_1)\in D$. Tällöin on olemassa $\delta > 0$ siten, että alkuarvotehtävällä
\[
\begin{cases}
y'' = f(x, y, y') \\
y(x_0) = y_0 \\
y'(x_0) = y_1
\end{cases}
\]
on olemassa välillä $]x_0 - \delta, x_0 + \delta[$ määritelty ratkaisu $y$. Lisäksi jos $y_1: \Delta_1 \to \R$ ja $y_2: \Delta_1 \to \R$ ovat alkuarvotehtävän ratkaisuja, niin $y_1(x) = y_2(x) \forall x \in \Delta_1 \cap \Delta_1$

\subsection{Ensimmäiseen kertalukuun palautuvia yhtälöitä}
\subsubsection{Muotoa $y'' = f(x, y')$ olevat yhtälöt}
\begin{example}
$y'' + \frac{1}{x}y' = \frac{1}{x}$, $x > 0$. Laitetaan $z = y'$

$z' + \frac{1}{x}z = \frac{1}{x}$, $x > 0$

$z' = \frac{1}{x} - \frac{1}{x}z = \frac{1}{x}(1 - z)$

Ensimmäisen kertaluvun separoituva yhtälö!
\end{example}

\subsubsection{Muotoa $y''=f(y, y')$ olevat yhtälöt}
Olkoon $y: \Delta \to \R$ yhtälön $y'' = f(y, y')$ ratkaisu, jolle $y'(x) \neq 0 \forall x \in \Delta$. Tällöin $y'(x) > 0$ tai $y'(x) < 0 \forall x \in \Delta$, joten $y$ on aidosti monotoninen. Siten on olemassa käänteisfunktio $y^{-1}: y(\Delta) \to \Delta$. Merkitään 
\[
z(t) = y'(y^{-1}(t)), t\in y(\Delta)
\]
\[
z'(t) = \frac{f(t, z)}{z}
\]

\begin{example}
Tarkastellaan yhtälöä $y''  \frac{1 + (y')^2}{y}$, $y \neq 0$

$z' = \frac{f(t, z)}{z} = \frac{\frac{1 + z^2}{t}}{z} = \frac{1 + z^2}{tz} = \frac{1}{t} \frac{1 + z^2}{z}$ Separoituva!

$z(t) = \pm \sqrt{(C_1t)^2 -1}$, $t \in \R \setminus [-1/|C_1|, 1/|C_1|], C_1 \neq 0$

$y'(x) = \pm \sqrt{(C_1 y(x))^2 -1}$

$\pm \int \frac{1}{\sqrt{(C_1y)^2 -1}} dy = \int 1 dx$
\end{example}

\subsection{Toisen kertaluvun lineaarinen yhtälö}
Määritelmä: Toisen kertaluvun normaalimuotoinen differentiaaliyhtälö on lineaarinen, jos se on muotoa
\[
y''(x) + p(x)y'(x) + r(x)y(x) = q(x)
\]
missä $p, q, r: \Delta \to \R$ ovat jatkuvia. Yhtälö on homogeeninen, jos $q(x) = 0$ ja vakiokertoiminen, jos $p$ ja $r$ ovat vakiofunktioita.

\subsubsection{Homogeeniyhtälön ratkaiseminen}
Määritelmä: Funktiot $u, v: \Delta \to \R$ ovat lineaarisesti riippuvia, jos on olemassa vakiot $c_1, c_2 \in \R$, joista ainakin toinen on nollasta eroava siten, että $c_1u(x) + c_2v(x) = 0$, $\forall x \in \Delta$. Muussa tapauksessa funktiot $u$ ja $v$ ovat lineaarisesti riippumattomia. 

Määritelmä: Jatkuvasti derivoituva funktioiden $y_1, y_2 : \Delta \to \R$ Wronskin determinantti on jatkuva funktio
\[
W = W(y_1, y_2): \Delta \to \R, W(x)  = \textrm{det}\begin{bmatrix}
y_1(x) & y_2(x) \\
y_1'(x) & y_2'(x) \\
\end{bmatrix} = y_1(x)y_2'(x) - y_2(x)y_1'(x)
\]

Lemma: Jos $y_1, y_2: \Delta \in \R$ ovat jatkuvasti derivoituvia ja $W(y_1, y_2)(x_0) \neq 0$ joillakin $x_0 \in \Delta$, niin $\{y_1, y_2 \}$ ovat lineaarisesti riippumattomia. 

Lause: Jos $y_1, y_2: \Delta \to \R$ ovat homogeeniyhtälön $y'' + p(x)y' + r(x)y = 0$ ratkaisuja, niin $\{y_1, y_2 \}$ on lineaarisesti riippumattomia, jos ja vain jos $W(y_1, y_2)(x_0) \neq 0$ jollakin $X_0 \in \Delta$. 

Seuraus: Jos $y_1, y_2: \Delta \to \R$ ovat homogeeniyhtälön $y'' + p(x)y' + r(x)y = 0$ lineaarisesti riippumattomia ratkaisuja, niin funktiolla $y_2$ on nollakohta funktion $y_1$ kahden nollakohdan välissä.

Määritelmä: Kahdesti jatkuvasti derivoituvat funktiot $y_1, y_2$ muodostavat homogeeniyhtälön $y'' + p(x)y' + r(x)y = 0$ ratkaisukannan, jos $y_1$ ja $y_2$ ovat yhtälön ratkaisuja ja jokaiselle ratkaisulle $y: \Delta \to \R$ on olemassa vakiot $c_1, c_2 \in \R$ siten, että $y(x) = c_1 y_1(x) + c_2 y_2(x)$, $\forall x \in  \Delta$. 

Lause: Seuraavat ehdot ovat yhtäpitäviä homogeeniyhtälön $y'' + p(x)y' + r(x)y = 0$ ratkaisuille $y_1$ ja $y_2$:
\begin{enumerate}
    \item Funktiot $y_1$ ja $y_2$ muodostavat yhtälön ratkaisukannan
    \item $\{y_1, y_2 \}$ ovat lineaarisesti riippumattomia
    \item $W(y_1, y_2)(x_0) \neq 0$ jollakin $x_0 \in \Delta$
    \item $W(y_1, y_2)(x) \neq 0$, $\forall x \in \Delta$
\end{enumerate}

\subsubsection{Ratkaisukannan löytäminen kertaluvun pudotuksella}
\begin{example}
$y'' - \frac{1}{x}y' + \frac{1}{x^2}y = 0$, $x \neq 0$, $y_1(x) = x$ Ensimmäinen ratkaisu

$y_2(x) = v(x)x$ ja $y_2'(x) = v'(x)x + v(x)$

$y2''(x) = v''(x)x + v'(x) + v'(x) = v''(x)x + 2v'(x)$

$y_2'' - \frac{1}{x}y_2' + \frac{1}{x^2}y_2 = v''x + 2v' - v' - \frac{1}{x}v + \frac{1}{x}v = v''x + v' = 0$ jos ja vain jos
$v'' = - \frac{1}{x}v'$. 

Separoituva yhtälö $v'(x) = 1/x$, $v(x) = \ln |x|$, $y_2(x) = x \ln |x|$. Koska $W(y_1, y_2)(x) = x$, $\Delta = ]-\infty, 0[$ ja $\Delta = ]0, \infty[$. 

$y(x) = c_1 x + c_2 x \ln|x|$, $c_1, c_2 \in \R$
\end{example}

\subsubsection{Vakiokertoiminen yhtälö}
$y'' + ay' + by = 0$, $a, b \in \R$ karakteristinen yhtälö $\lambda^2 + a\lambda + b = 0$ ja $\lambda = - \frac{a}{2} \pm \sqrt{\frac{a^2}{4} - b}$

Lause: Vakiokertoimisen homogeeniyhtälön $y'' + ay + by = 0$ kaikki ratkaisut saadaam muodossa
\begin{enumerate}
    \item $y(x) = C_1e^{\lambda_1 x} + C_2 e^{\lambda_2 x}$, jos $\lambda_1, \lambda_2 \in \R$ ovat karakteristisen yhtälön juuria.
    \item $y(x) = C_1e^{\lambda x} + C_2xe^{\lambda x}$ jos $\lambda \in \R$ on karakteristisen yhtälön karakteristinen juuri
    \item $y(x) = c_1 e^{\alpha x}\sin (\beta x) + c_2 e^{\alpha x} \cos (\beta x)$, jos $\alpha \pm i \beta \in \C$ ovat karakteristisen yhtälön kompleksijuuria
\end{enumerate}

\begin{example}
\[
\begin{cases}
y'' + 2y' + 2y = 0 \\
y(0) = 0 \\
y'(0) = 1 \\
\end{cases}
\]
$\lambda^2 + 2\lambda + 2 = 0$, $\lambda = -1 \pm i$

$y(x) = C_1e^{-x}\sin x + C_2 e^{-x}\cos x$, $C_1, C_2 \in \R$

$C_2 = 0$, $y'(x) = -C-1 e^{-x}\sin x + C-1 e^{-x} \cos x$, $y'(0) = 1$, $C_1 = 1$, $y(x) = e^{-x} \sin x$
\end{example}

\subsubsection{Yleisen lineaariyhtälön ratkaiseminen}

\end{example}

$y_3(x) = C_1(x)y_1(x) + C_2(x)y_2(x)$

\section{Yleistä teoriaa}
\subsection{Lipschitz-jatkuvuus ja aputuloksia}
Ensimmäisen kertaluvun normaalimuotoinen differentiaaliyhtälö alkuarvotehtävä, $y'(x) = f(x, y(x))$, $y(x_0) = y_0$ ratkaisun olemassaolo ja yksikäsitteisyys

Määritelmä: Olkoon $I$ väli $\R$:ssä. Funktio $h: I \to \R$ on Lipschitz-jatkuva välillä $I$, jos on olemassa sellainen vakio $M \geq 0$, että
\[
|h(x) - h(y) | \leq M|x-y|, \forall x, y \in I
\]
Olkoon $H$ tason $R^2$ osajoukko. Funktio $f:H \to \R$ on $H$:ssa tasaisesti Lipschitz-jatkuva muuttujan $y$ suhteen, jos on olemassa sellainen vakio $M \geq 0$, että
\[

\subsection{Kriittinen piste, stabiilisuus ja lineaarinen tapaus}
$z(t) = (x(t), y(t)), I \subset \R$

Määritelmä: Piste $z_0 = (x_0, y_0) \in \R^2$ on autonomisen systeemin kriittinen piste, jos $f(x_0, y_0) = y(x_0, y_0) = 0$. Niiden joukko on AS:n kriittinen joukko. Kriittistä pistettä $(x_0, y_0)$ vastaa AS:n vakioratkaisu $x(t) \equiv x_0, y(t) \equiv y_0$ ja kääntäen. Tätä vakioratkaisua kutsutaan AS:n tasapainotilaksi. 

Kriittinen piste $z_0$ (tasapainotila) on stabiili, jos jokaista $\epsilon > 0$ kohti on olemassa sellainen $\delta > 0$, että systeemin ratkaisulle $z(t) = (x(t), y(t))$ pätee
\[
\| z(t) - z_0 \| < \epsilon \forall t \geq 0
\]
ainakin $\|z(0) - z_0\| < \delta$. Muussa tapauksessa piste $z_0$ on epästabiili. 

Jos kriittinen piste $z_0 = (x_0, y_0)$ on stabiili, ja lisäksi on olemassa sellainen $\eta > 0$, että ratkaisulle $z(t) = (x(t), y(t))$ pätee
\[
\lim_{t \to \infty} z(t) = z_0 \iff \lim_{t \to \infty}x(t) = x_0 \lor \lim_{t \to \infty}y(t) = y_0
\]
aina kun $\|z(0) - z_0\| < n$, niin $z_0$ on asymptoottisesti stabiili.

\subsection{Kriittisen pisteen siirto origoon}
\begin{example}
\[\begin{cases}

Kriittinen piste $(x_0, y_0) = (1,2)$, $-5x + 2y + 1 = x - 4y + 7 = 0$

$A = \begin{bmatrix}-5 & 2 \\1 & -4\end{bmatrix}$

$\lambda_1 = -3$, $\lambda_2 = -6$

\[

\]
$t \to \infty$m asymptoottisesti stabiili!
\end{example}

Lause: Lineaarisen autonomisen systeemin stabiilisuus
\[
\dot{z}(t) = Az(t), z=(x, y), A = \begin{bmatrix}a & b \\c & d\end{bmatrix} \in \R^{2 \times 2}
\]
kriittisen pisteen $0$ laatu määräytyy matriisin $A$ ominaisarvoista seuraavasti: Jos pätee $\textrm{det}A \neq 0$, niin 0 on systeemin ainoa kriittinen piste. Jos $\textrm{det}(A) = 0$, kriittisiä pisteitä on $\infty$ monta, ja niiden laatu on sama kuin kriittisen pisteen $0$ eli saadaan alla olevasta taulukosta:

\subsection{Epälineaariset autonomiset systeemit tasossa}
Lause: Jos $z(t) = (x(t), y(t))$ on parin ratkaisu välillä $I$ ja $a \in \R$ on mielivaltainen vakio, niin myös funktio $z_{\alpha}: I - a \to \R^2$, $z_{\alpha}(t) = z(t + a), t \in I - a$ on parin ratkaisu. 

Lause: Olkoot parin määrittelemät funktiot $f$ ja $g$ jatkuvia, ja olkoon parilla ratkaisu $z(t) = (x(t), y(t))$ välillä $[a, \infty[ (a \in \R)$. Jos on olemassa raja-arvo
\[
z_{\infty} = \lim_{t \to \infty}z(t) = (\lim_{t \to \infty}x(t), \lim_{t \to \infty}y(t)) = (x_{\infty}, y_{\infty}) \in \R^2
\]
niin $z_{\infty}=(x_{\infty}, y_{\infty})$ on parin kriittinen piste.

Lause: (Poincarén stabiilisuslause). Oletetaan että autonomisessa systeemissä origo $0$ on kriittinen piste, ja funktiot $f(x, y)$ sekä $g(x, y)$ ovat jatkuvasti derivoituvia jossain sen ympäristössä. Olkoot $\lambda_1$ ja $\lambda_2$ origon linearisointiin liittyvän derivaattamatriisin $A$ ominaisarvot. Oletetaan vielä että $\lambda_1 \neq \lambda_2$ ja $\textrm{det}(A) \neq 0$. Tällöin systeemissä kriittisen pisteen $0$ laatu on sama kuin se linearisoinnissa, poikkeuksena yksi tapaus: jos $\lambda_1$ ja $\lambda_2$ ovat puhtaasti imaginaarisia, laatua ei voi päätellä linearisoinnista.

     Erimerkkiset & Epästabiili & Epästabiili \\
     Reaaliosa pos. & Epästabiili & Epästabiili \\
     Reaaliosa neg. & Asymptoottisesti stabiili & Epästabiili \\

\begin{example}
\[
\begin{cases}
\dot{x}(t) = 7x - x^2 - 2xy \\
\dot{y}(t) = 5y -y^2 -xy \\
\end{cases}
\]
$x(7-x-2y) = 0$, $y(5 - y -x) = 0$, $(x, y) = (0,0)$, $(x, y) = (0,5)$, $(x,y) = (7, 0)$, $(x, y) = (3,2)$

\[
A(x, y) = \begin{bmatrix}\frac{\partial f(x, y)}{\partial x} & \frac{\partial f(x, y)}{\partial y} \\ \frac{\partial g(x, y)}{\partial x} & \frac{\partial g(x, y)}{\partial y}\end{bmatrix} = \begin{bmatrix}7 -2x -2y & -2x \\ -y & 5-2y-x \end{bmatrix}

$(x, y) = (0,0)$, $A = \begin{bmatrix}7 & 0 \\ 0 & 5\end{bmatrix}$, $\lambda_1 = 5 \land \lambda_2 = 7$ $\to$ epästabiili
\end{example}
